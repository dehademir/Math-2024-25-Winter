```
import numpy as np
import matplotlib.pyplot as plt

a = np.array([3, 4])
length_a = np.linalg.norm(a)
k = 1 / length_a

b = np.array([1, 1])
length_b = np.linalg.norm(b)
unit_b = b / length_b

plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label="Vector b")
plt.quiver(0, 0, unit_b[0], unit_b[1], angles='xy', scale_units='xy', scale=1, color='r', label="Unit vector of b")
plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.legend()
plt.show()

c = np.array([1, 2, 3])
length_c = np.linalg.norm(c)
unit_c = c / length_c

v = np.array([2, 3, 4])
b1 = np.array([1, 0, 1])
b2 = np.array([0, 1, 0])
b3 = np.array([1, 0, -1])

A = np.array([b1, b2, b3]).T
v_cartesian = np.linalg.solve(A, v)

print(f"{k}")
print(f"{length_b}")
print(f"{unit_b}")
print(f"{length_c}")
print(f"{unit_c}")
print(f"{v_cartesian}")
```
