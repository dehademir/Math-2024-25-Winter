```
import numpy as np
import matplotlib.pyplot as plt

a = np.array([2, 1])
b = np.array([-1, 1])
sum_ab = a + b

area = 0.5 * np.abs(np.cross(a, b))

a_3d = np.array([2, 1, 0])
b_3d = np.array([-1, 1, 0])
c_3d = np.array([1, 2, 0])
cross_bc = np.cross(b_3d, c_3d)
volume = np.abs(np.dot(a_3d, cross_bc))

dot_product_ab = np.dot(a, b)
perpendicular = dot_product_ab == 0

a_3d_angle = np.array([4, 2, 1])
b_3d_angle = np.array([1, 3, 2])
cos_theta = np.dot(a_3d_angle, b_3d_angle) / (np.linalg.norm(a_3d_angle) * np.linalg.norm(b_3d_angle))
angle_deg = np.degrees(np.arccos(cos_theta))

cross_bc = np.cross(b_3d, c_3d)
lhs = np.cross(a_3d, cross_bc)
rhs = (np.dot(a_3d, c_3d) * b_3d) - (np.dot(a_3d, b_3d) * c_3d)

plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='b', label="Vector a")
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='g', label="Vector b")
plt.quiver(0, 0, sum_ab[0], sum_ab[1], angles='xy', scale_units='xy', scale=1, color='r', label="Sum of a and b")
plt.xlim(-2, 3)
plt.ylim(-2, 3)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.legend()
plt.show()

print(f"Sum of vectors: {sum_ab}")
print(f"Area of triangle: {area}")
print(f"Volume of parallelepiped: {volume}")
print(f"Are vectors perpendicular? {perpendicular}")
print(f"Angle between vectors in degrees: {angle_deg}")
print(f"Left-hand side of identity: {lhs}")
print(f"Right-hand side of identity: {rhs}")
```
