```
import sympy as sp

x, t = sp.symbols('x t')

y1 = -3*x + 3
y1_prime = sp.diff(y1, x)

y2 = sp.pi*x + sp.sin(1)
y2_prime = sp.diff(y2, x)

y3 = 4 + sp.sin(2)
y3_prime = sp.diff(y3, x)

y4 = 2*x**3 - 3*x**2 + 8*x - 9
y4_prime = sp.diff(y4, x)

y5 = 6*x**(1/3)
y5_prime = sp.diff(y5, x)

y6 = x
y6_prime = sp.diff(y6, x)

y7 = sp.cos(x) + sp.sin(x)
y7_prime = sp.diff(y7, x)

y8 = 2*sp.sin(x)*sp.cos(x)
y8_prime = sp.diff(y8, x)

y9 = x*sp.sin(x)
y9_prime = sp.diff(y9, x)

y10 = (x + 1)*(x + 1)
y10_prime = sp.diff(y10, x)

y11 = x*(x + 1)
y11_prime = sp.diff(y11, x)

y12 = (x + 1)*sp.exp(x)
y12_prime = sp.diff(y12, x)

y13 = sp.sin(x**2)
y13_prime = sp.diff(y13, x)

y14 = sp.exp(-2*x)
y14_prime = sp.diff(y14, x)

y15 = 1/sp.sin(x + 1)
y15_prime = sp.diff(y15, x)

y16 = 2*x + 1
y16_prime = sp.diff(y16, x)

ln_sin_x = sp.ln(sp.sin(x))
ln_sin_x_prime = sp.diff(ln_sin_x, x)

f_x = sp.cos(x)
f_x_prime = sp.diff(f_x, x)
f_x_double_prime = sp.diff(f_x_prime, x)

lim1 = sp.limit(sp.sin(x)/x, x, 0)

lim2 = sp.limit(sp.ln(x)/x, x, sp.oo)

lim3 = sp.limit(sp.exp(x)/x, x, sp.oo)

x_t = 3*t**2 - 6*t + 1
V_t = sp.diff(x_t, t)
a_t = sp.diff(V_t, t)

V_at_2 = V_t.subs(t, 2)
a_at_2 = a_t.subs(t, 2)

print(f"Derivative of y1(x) = -3x + 3: {y1_prime}")
print(f"Derivative of y2(x) = πx + sin(1): {y2_prime}")
print(f"Derivative of y3(x) = 4 + sin(2): {y3_prime}")
print(f"Derivative of y4(x) = 2x^3 - 3x^2 + 8x - 9: {y4_prime}")
print(f"Derivative of y5(x) = 6x^(1/3): {y5_prime}")
print(f"Derivative of y6(x) = x: {y6_prime}")
print(f"Derivative of y7(x) = cos(x) + sin(x): {y7_prime}")
print(f"Derivative of y8(x) = 2sin(x)cos(x): {y8_prime}")
print(f"Derivative of y9(x) = x*sin(x): {y9_prime}")
print(f"Derivative of y10(x) = (x+1)(x+1): {y10_prime}")
print(f"Derivative of y11(x) = x(x + 1): {y11_prime}")
print(f"Derivative of y12(x) = (x + 1)*exp(x): {y12_prime}")
print(f"Derivative of y13(x) = sin(x^2): {y13_prime}")
print(f"Derivative of y14(x) = exp(-2x): {y14_prime}")
print(f"Derivative of y15(x) = 1/sin(x + 1): {y15_prime}")
print(f"Derivative of y16(x) = 2x + 1: {y16_prime}")
print(f"Derivative of ln(sin(x)): {ln_sin_x_prime}")
print(f"f''(x) for f(x) = cos(x): {f_x_double_prime}")
print(f"Limit (x -> 0) sin(x)/x: {lim1}")
print(f"Limit (x -> ∞) ln(x)/x: {lim2}")
print(f"Limit (x -> ∞) exp(x)/x: {lim3}")
print(f"Velocity at t = 2: {V_at_2}")
print(f"Acceleration at t = 2: {a_at_2}")
```
