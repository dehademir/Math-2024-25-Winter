# Equations of Lines in a Plane

## 1. **General Form of a Line**
The general equation of a line in a 2D plane is given by:

$Ax + By + C = 0$

Where:
- $A$ and $B$ are coefficients that determine the slope and direction of the line,
- $C$ is a constant term.

### Example:
- $2x + 3y - 6 = 0$ represents a line with slope $-2/3$ and y-intercept $2$.

## 2. **Slope-Intercept Form**
The slope-intercept form of the equation of a line is:

$y = mx + b$

Where:
- $m$ is the slope of the line,
- $b$ is the y-intercept (the point where the line intersects the y-axis).

### Example:
- $y = 2x + 3$ represents a line with slope $2$ and y-intercept $3$.

## 3. **Point-Slope Form**
The point-slope form of the equation of a line is:

$y - y_1 = m(x - x_1)$

Where:
- $(x_1, y_1)$ is a point on the line,
- $m$ is the slope of the line.

### Example:
- For the point $(1, 2)$ and slope $3$, the equation of the line is $y - 2 = 3(x - 1)$.

## 4. **Standard Form**
The standard form of the equation of a line is a rearranged version of the general form, where $A$, $B$, and $C$ are integers, and $A \geq 0$.

$Ax + By = C$

### Example:
- $2x + 3y = 6$ is in standard form.

## 5. **Intercept Form**
The intercept form of the equation of a line is:

$\frac{x}{a} + \frac{y}{b} = 1$

Where:
- $a$ is the x-intercept (the point where the line intersects the x-axis),
- $b$ is the y-intercept (the point where the line intersects the y-axis).

### Example:
- $\frac{x}{2} + \frac{y}{3} = 1$ represents a line with x-intercept $2$ and y-intercept $3$.

# Equation of the Line Passing Through Points A(1, 2) and B(3, 4)

### Step 1: Calculate the Slope
The slope $m$ of a line passing through two points $(x_1, y_1)$ and $(x_2, y_2)$ is given by:

$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$

For points $A(1, 2)$ and $B(3, 4)$:

$$
m = \frac{4 - 2}{3 - 1} = \frac{2}{2} = 1
$$

### Step 2: Use the Point-Slope Form
The point-slope form of the equation of a line is:

$$
y - y_1 = m(x - x_1)
$$

Using point $A(1, 2)$ and the slope $m = 1$:

$$
y - 2 = 1(x - 1)
$$

Simplifying:

$$
y - 2 = x - 1
$$

$$
y = x + 1
$$

### Final Answer:
The equation of the line passing through points $A(1, 2)$ and $B(3, 4)$ is:

$$
y = x + 1
$$


```
import numpy as np
import matplotlib.pyplot as plt

# Coordinates of points A and B
A = (1, 2)
B = (3, 4)

# Calculate the slope (m) of the line
m = (B[1] - A[1]) / (B[0] - A[0])

# Equation of the line: y - y1 = m(x - x1)
# Using point A(1, 2)
x_vals = np.linspace(-1, 5, 400)
y_vals = m * (x_vals - A[0]) + A[1]

# Create the plot
plt.figure(figsize=(6, 6))

# Plot the line
plt.plot(x_vals, y_vals, label=r'$y = x + 1$', color='b')

# Plot the points A and B
plt.scatter([A[0], B[0]], [A[1], B[1]], color='r', zorder=5)
plt.text(A[0], A[1], 'A(1, 2)', fontsize=12, verticalalignment='bottom', horizontalalignment='right')
plt.text(B[0], B[1], 'B(3, 4)', fontsize=12, verticalalignment='top', horizontalalignment='left')

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Visualization of the Line Passing Through A(1, 2) and B(3, 4)')

# Set equal scaling for both axes
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

### Equation of the Line Passing Through Point A(1, 2) and Parallel to $ y = 2x + 3 $

The equation of a line parallel to another line has the same slope. The given line is $ y = 2x + 3 $, so the slope of the new line will also be $ 2 $.

To find the equation of the line passing through $ A(1, 2) $ and parallel to $ y = 2x + 3 $, we will use the point-slope form of the equation:

$$
y - y_1 = m(x - x_1)
$$

Where:
- $ m $ is the slope of the line,
- $ (x_1, y_1) $ is the given point on the line.

Using point $ A(1, 2) $ and slope $ m = 2 $, we get:

$$
y - 2 = 2(x - 1)
$$

Simplifying:

$$
y - 2 = 2x - 2
$$

$$
y = 2x
$$

### Final Answer:
The equation of the line passing through point $ A(1, 2) $ and parallel to $ y = 2x + 3 $ is:

$$
y = 2x
$$


```
import numpy as np
import matplotlib.pyplot as plt

# Coordinates of point A
A = (1, 2)

# Slope of the given line y = 2x + 3
m = 2

# Equation of the line: y - y1 = m(x - x1)
# Using point A(1, 2)
x_vals = np.linspace(-1, 5, 400)
y_vals = m * (x_vals - A[0]) + A[1]

# Create the plot
plt.figure(figsize=(6, 6))

# Plot the line
plt.plot(x_vals, y_vals, label=r'$y = 2x$', color='b')

# Plot the point A
plt.scatter(A[0], A[1], color='r', zorder=5)
plt.text(A[0], A[1], 'A(1, 2)', fontsize=12, verticalalignment='bottom', horizontalalignment='right')

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Visualization of the Line Passing Through A(1, 2) and Parallel to y = 2x + 3')

# Set equal scaling for both axes
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

### Equation of the Line Passing Through Point A(1, 2) and Perpendicular to $ y = 2x + 3 $

To find the equation of the line that passes through point $ A(1, 2) $ and is perpendicular to the line $ y = 2x + 3 $, we need to use the fact that the slopes of two perpendicular lines are negative reciprocals of each other.

The slope of the given line $ y = 2x + 3 $ is $ 2 $. The slope of the line perpendicular to it will be the negative reciprocal of $ 2 $, which is:

$$
m_{\text{perpendicular}} = -\frac{1}{2}
$$

Now we can use the point-slope form of the equation:

$$
y - y_1 = m(x - x_1)
$$

Where:
- $ m_{\text{perpendicular}} = -\frac{1}{2} $,
- $ (x_1, y_1) = (1, 2) $.

Substitute the values:

$$
y - 2 = -\frac{1}{2}(x - 1)
$$

Simplify:

$$
y - 2 = -\frac{1}{2}x + \frac{1}{2}
$$

$$
y = -\frac{1}{2}x + \frac{5}{2}
$$

### Final Answer:
The equation of the line passing through point $ A(1, 2) $ and perpendicular to $ y = 2x + 3 $ is:

$$
y = -\frac{1}{2}x + \frac{5}{2}
$$


```
import numpy as np
import matplotlib.pyplot as plt

# Coordinates of point A
A = (1, 2)

# Slope of the given line y = 2x + 3
m_perpendicular = -1/2

# Equation of the line: y - y1 = m(x - x1)
# Using point A(1, 2)
x_vals = np.linspace(-1, 5, 400)
y_vals = m_perpendicular * (x_vals - A[0]) + A[1]

# Create the plot
plt.figure(figsize=(6, 6))

# Plot the perpendicular line
plt.plot(x_vals, y_vals, label=r'$y = -\frac{1}{2}x + \frac{5}{2}$', color='b')

# Plot the point A
plt.scatter(A[0], A[1], color='r', zorder=5)
plt.text(A[0], A[1], 'A(1, 2)', fontsize=12, verticalalignment='bottom', horizontalalignment='right')

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Visualization of the Perpendicular Line Passing Through A(1, 2)')

# Set equal scaling for both axes
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

### Intersection Point and Angle Between the Lines $ y = 2x + 3 $ and $ y = 3x + 2 $

#### Finding the Intersection Point

We are given two lines:

1. $ y = 2x + 3 $
2. $ y = 3x + 2 $

To find the intersection point, set the right-hand sides of the equations equal to each other:

$$
2x + 3 = 3x + 2
$$

Solving for $ x $:

$$
2x - 3x = 2 - 3
$$

$$
-x = -1
$$

$$
x = 1
$$

Now substitute $ x = 1 $ into one of the original equations (e.g., $ y = 2x + 3 $):

$$
y = 2(1) + 3 = 5
$$

Thus, the intersection point is $ (1, 5) $.

#### Finding the Angle Between the Lines

The formula for the angle $ \theta $ between two lines with slopes $ m_1 $ and $ m_2 $ is:

$$
\tan(\theta) = \left| \frac{m_1 - m_2}{1 + m_1 m_2} \right|
$$

For the given lines, the slopes are:

- $ m_1 = 2 $ (from $ y = 2x + 3 $)
- $ m_2 = 3 $ (from $ y = 3x + 2 $)

Substituting into the formula:

$$
\tan(\theta) = \left| \frac{2 - 3}{1 + 2 \times 3} \right| = \left| \frac{-1}{7} \right| = \frac{1}{7}
$$

Thus, the angle $ \theta $ is:

$$
\theta = \tan^{-1}\left( \frac{1}{7} \right)
$$

This is the angle between the two lines.

### Final Answer:

- The intersection point of the lines is $ (1, 5) $.
- The angle between the lines is:

$$
\theta = \tan^{-1}\left( \frac{1}{7} \right)
$$


```
import numpy as np
import matplotlib.pyplot as plt

# Define the lines y = 2x + 3 and y = 3x + 2
def line1(x):
    return 2*x + 3

def line2(x):
    return 3*x + 2

# Intersection point (1, 5)
intersection = (1, 5)

# Create x values for plotting
x_vals = np.linspace(-2, 4, 400)

# Plot the lines
plt.plot(x_vals, line1(x_vals), label=r'$y = 2x + 3$', color='b')
plt.plot(x_vals, line2(x_vals), label=r'$y = 3x + 2$', color='g')

# Plot the intersection point
plt.scatter(intersection[0], intersection[1], color='r', zorder=5)
plt.text(intersection[0], intersection[1], 'Intersection (1, 5)', fontsize=12, verticalalignment='bottom', horizontalalignment='right')

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Intersection of the Lines $y = 2x + 3$ and $y = 3x + 2$')

# Set equal scaling for both axes
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

### Equation of the Line Passing Through Point $ A(1, 2) $ and Parallel to Vector $ \mathbf{v} = [2, 3] $

We are given a point $ A(1, 2) $ and a vector $ \mathbf{v} = [2, 3] $. The parametric equation of the line passing through this point and parallel to the vector is given by:

$$
\mathbf{r}(t) = \mathbf{r}_0 + t \mathbf{v}
$$

Where:

- $ \mathbf{r}_0 = (1, 2) $ is the point through which the line passes.
- $ \mathbf{v} = [2, 3] $ is the direction vector.
- $ t $ is the parameter.

In terms of coordinates, the parametric equations for $ x $ and $ y $ are:

$$
x(t) = 1 + 2t
$$

$$
y(t) = 2 + 3t
$$

Alternatively, we can write the equation of the line in slope-intercept form. The slope $ m $ of the line is given by:

$$
m = \frac{3}{2}
$$

Using point-slope form, the equation of the line is:

$$
y - 2 = \frac{3}{2}(x - 1)
$$

Simplifying:

$$
y = \frac{3}{2}(x - 1) + 2
$$

This is the equation of the line.


```
import numpy as np
import matplotlib.pyplot as plt

# Define the parametric equations of the line
def line(t):
    x = 1 + 2 * t
    y = 2 + 3 * t
    return x, y

# Generate values for t
t_values = np.linspace(-10, 10, 400)

# Compute x and y values of the line
x_values, y_values = line(t_values)

# Plot the line
plt.figure(figsize=(6,6))
plt.plot(x_values, y_values, label='Line passing through A(1, 2) parallel to v=[2, 3]')

# Plot point A
plt.scatter(1, 2, color='red', zorder=5)
plt.text(1.1, 2, 'A(1, 2)', fontsize=12)

# Set up plot limits and labels
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Line through point A(1, 2) parallel to vector v = [2, 3]')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

#Example-6

### Example of a Line Parallel and Perpendicular to the Line $ y = 2x + 3 $

#### Parallel Line

A line parallel to $ y = 2x + 3 $ will have the same slope. The slope of the line $ y = 2x + 3 $ is $ 2 $. Therefore, a line parallel to this one could have the equation:

$$
y = 2x + 1
$$

This is a line that has the same slope but a different $ y $-intercept.

#### Perpendicular Line

The slope of a line perpendicular to $ y = 2x + 3 $ is the negative reciprocal of $ 2 $. The negative reciprocal of $ 2 $ is $ -\frac{1}{2} $. Therefore, a line perpendicular to $ y = 2x + 3 $ could have the equation:

$$
y = -\frac{1}{2}x + 4
$$

This line has a slope of $ -\frac{1}{2} $ and a different $ y $-intercept.


```
import numpy as np
import matplotlib.pyplot as plt

# Define the line equations
def original_line(x):
    return 2 * x + 3

def parallel_line(x):
    return 2 * x + 1

def perpendicular_line(x):
    return -0.5 * x + 4

# Generate x values
x_values = np.linspace(-10, 10, 400)

# Calculate y values for each line
y_original = original_line(x_values)
y_parallel = parallel_line(x_values)
y_perpendicular = perpendicular_line(x_values)

# Plot the lines
plt.figure(figsize=(6,6))
plt.plot(x_values, y_original, label='$y = 2x + 3$ (Original Line)', color='blue')
plt.plot(x_values, y_parallel, label='$y = 2x + 1$ (Parallel Line)', color='green', linestyle='--')
plt.plot(x_values, y_perpendicular, label='$y = -\\frac{1}{2}x + 4$ (Perpendicular Line)', color='red', linestyle=':')

# Set up plot limits and labels
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim(-10, 10)
plt.ylim(-10, 20)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Parallel and Perpendicular Lines to $y = 2x + 3$')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

### Distance from Point $ A(1, 2) $ to the Line $ y = 2x + 3 $

To find the distance between a point and a line, we can use the following formula:

$$
d = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}}
$$

where the equation of the line is given in the form $ Ax + By + C = 0 $, and $ (x_1, y_1) $ is the point from which we want to calculate the distance.

#### Step 1: Convert the equation of the line to the form $ Ax + By + C = 0 $

The given equation of the line is:

$$
y = 2x + 3
$$

Rearrange it to the form $ Ax + By + C = 0 $:

$$
2x - y + 3 = 0
$$

Thus, $ A = 2 $, $ B = -1 $, and $ C = 3 $.

#### Step 2: Substitute the values into the distance formula

We are given the point $ A(1, 2) $, so $ x_1 = 1 $ and $ y_1 = 2 $. Now, we can calculate the distance:

$$
d = \frac{|2(1) - (2) + 3|}{\sqrt{2^2 + (-1)^2}} = \frac{|2 - 2 + 3|}{\sqrt{4 + 1}} = \frac{|3|}{\sqrt{5}} = \frac{3}{\sqrt{5}}
$$

Thus, the distance from point $ A(1, 2) $ to the line is:

$$
d = \frac{3}{\sqrt{5}} \approx 1.34
$$


```
import numpy as np
import matplotlib.pyplot as plt

# Given values
A_x, A_y = 1, 2  # Point A
A, B, C = 2, -1, 3  # Coefficients from the line equation 2x - y + 3 = 0

# Define the line equation
def line_eq(x):
    return 2 * x + 3

# Generate x values
x_values = np.linspace(-5, 5, 400)
y_values = line_eq(x_values)

# Plot the line and point A
plt.figure(figsize=(6, 6))
plt.plot(x_values, y_values, label='$y = 2x + 3$', color='blue')
plt.scatter(A_x, A_y, color='red', label=f'Point A({A_x}, {A_y})')

# Add the perpendicular line from the point to the line
slope_perpendicular = -1 / 2  # Perpendicular slope to the given line
y_intercept_perpendicular = A_y - slope_perpendicular * A_x  # Calculate y-intercept for perpendicular line
y_values_perpendicular = slope_perpendicular * x_values + y_intercept_perpendicular

plt.plot(x_values, y_values_perpendicular, label='Perpendicular from A', color='green', linestyle='--')

# Mark the intersection point of the perpendicular line and the original line
x_intersection = (A_y - 3 + 2 * A_x) / 5  # Calculate the x-coordinate of the intersection point
y_intersection = line_eq(x_intersection)

plt.scatter(x_intersection, y_intersection, color='purple', label=f'Intersection ({x_intersection:.2f}, {y_intersection:.2f})')

# Set up plot limits and labels
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim(-5, 5)
plt.ylim(-5, 15)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Distance from Point A to the Line $y = 2x + 3$')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

### Angle Between the Line $ y = x + 3 $ and the $ O_x $ Axis

The equation of the line is given by:

$$
y = x + 3
$$

This is a straight line with slope $ m = 1 $, where the general form of the equation of a line is:

$$
y = mx + b
$$

The angle $ \theta $ between the line and the $ O_x $ axis can be calculated using the following formula:

$$
\tan(\theta) = |m|
$$

Where $ m $ is the slope of the line.

#### Step 1: Calculate the Angle

For the given line, the slope $ m = 1 $. Therefore:

$$
\tan(\theta) = |1| = 1
$$

Now, we compute the angle:

$$
\theta = \tan^{-1}(1)
$$

Using the inverse tangent function:

$$
\theta = 45^\circ
$$

Thus, the angle between the line and the $ O_x $ axis is $ 45^\circ $.


```
import numpy as np
import matplotlib.pyplot as plt

# Define the line equation y = x + 3
def line_eq(x):
    return x + 3

# Generate x values for the plot
x_values = np.linspace(-5, 5, 400)
y_values = line_eq(x_values)

# Plot the line and the O_x axis
plt.figure(figsize=(6, 6))
plt.plot(x_values, y_values, label='$y = x + 3$', color='blue')

# Plot the O_x axis (x-axis)
plt.axhline(0, color='black',linewidth=1)

# Set up plot limits and labels
plt.xlim(-5, 5)
plt.ylim(-5, 10)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Angle Between Line and $ O_x $ Axis')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

### Finding a Vector Perpendicular to the Line $ x + y + 1 = 0 $

The equation of the line is given by:

$$
x + y + 1 = 0
$$

This equation can be written in the general form of a line:

$$
Ax + By + C = 0
$$

Here, $ A = 1 $ and $ B = 1 $.

The vector perpendicular to the line is given by the vector formed by the coefficients of $ x $ and $ y $, i.e., the vector:

$$
\mathbf{v} = \langle A, B \rangle
$$

For the line $ x + y + 1 = 0 $, the vector perpendicular to it is:

$$
\mathbf{v} = \langle 1, 1 \rangle
$$

Thus, a vector perpendicular to the line is $ \langle 1, 1 \rangle $.


```
import numpy as np
import matplotlib.pyplot as plt

# Define the line equation x + y + 1 = 0, rearranged as y = -x - 1
def line_eq(x):
    return -x - 1

# Generate x values for the plot
x_values = np.linspace(-5, 5, 400)
y_values = line_eq(x_values)

# Define the perpendicular vector
perpendicular_vector = np.array([1, 1])

# Plot the line and the perpendicular vector
plt.figure(figsize=(6, 6))
plt.plot(x_values, y_values, label='$x + y + 1 = 0$', color='blue')

# Plot the perpendicular vector
plt.quiver(0, 0, perpendicular_vector[0], perpendicular_vector[1], angles='xy', scale_units='xy', scale=1, color='red', label='Perpendicular Vector: (1, 1)')

# Set up plot limits and labels
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Perpendicular Vector to the Line $x + y + 1 = 0$')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
