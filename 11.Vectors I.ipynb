## Vectors

A **vector** is a mathematical entity that has both magnitude and direction. It is often represented as an ordered pair or triplet of numbers, which correspond to coordinates in space. Vectors are used in many fields, such as physics, engineering, and computer science, to represent quantities like velocity, displacement, and force.

### Components of a Vector

A vector in $ n $-dimensional space is represented as:

$$
\mathbf{v} = \langle v_1, v_2, \dots, v_n \rangle
$$

Where:
- $ \mathbf{v} $ is the vector.
- $ v_1, v_2, \dots, v_n $ are the components of the vector along each axis.

For example, in 2D space, a vector can be represented as:

$$
\mathbf{v} = \langle 3, 4 \rangle
$$

### Types of Vectors

1. **Zero Vector**: A vector whose components are all zeros, such as $ \mathbf{0} = \langle 0, 0 \rangle $. It has no direction and its magnitude is zero.

2. **Unit Vector**: A vector with a magnitude of 1. It is used to specify direction. Unit vectors are often denoted with a hat, such as $ \hat{i}, \hat{j}, \hat{k} $ in 3D space.

3. **Position Vector**: A vector that represents the position of a point in space, usually given in terms of coordinates.

4. **Direction Vector**: A vector that indicates the direction of a line or a motion.

### Vector Operations

#### 1. **Addition of Vectors**

The sum of two vectors $ \mathbf{v}_1 = \langle v_{1x}, v_{1y} \rangle $ and $ \mathbf{v}_2 = \langle v_{2x}, v_{2y} \rangle $ is obtained by adding their corresponding components:

$$
\mathbf{v}_1 + \mathbf{v}_2 = \langle v_{1x} + v_{2x}, v_{1y} + v_{2y} \rangle
$$

#### 2. **Scalar Multiplication**

Multiplying a vector by a scalar $ k $ scales the vector by a factor of $ k $:

$$
k \cdot \mathbf{v} = \langle k \cdot v_1, k \cdot v_2, \dots, k \cdot v_n \rangle
$$

#### 3. **Dot Product**

The **dot product** of two vectors $ \mathbf{v}_1 = \langle v_{1x}, v_{1y} \rangle $ and $ \mathbf{v}_2 = \langle v_{2x}, v_{2y} \rangle $ is calculated as:

$$
\mathbf{v}_1 \cdot \mathbf{v}_2 = v_{1x} \cdot v_{2x} + v_{1y} \cdot v_{2y}
$$

This operation results in a scalar and is used to find the angle between two vectors or determine if two vectors are orthogonal (dot product = 0).

#### 4. **Cross Product (in 3D)**

The **cross product** of two vectors $ \mathbf{v}_1 = \langle v_{1x}, v_{1y}, v_{1z} \rangle $ and $ \mathbf{v}_2 = \langle v_{2x}, v_{2y}, v_{2z} \rangle $ in 3D is calculated as:

$$
\mathbf{v}_1 \times \mathbf{v}_2 = \langle (v_{1y} \cdot v_{2z} - v_{1z} \cdot v_{2y}), (v_{1z} \cdot v_{2x} - v_{1x} \cdot v_{2z}), (v_{1x} \cdot v_{2y} - v_{1y} \cdot v_{2x}) \rangle
$$

The cross product results in a vector that is perpendicular to both $ \mathbf{v}_1 $ and $ \mathbf{v}_2 $.

#### 5. **Magnitude of a Vector**

The magnitude (length) of a vector $ \mathbf{v} = \langle v_1, v_2 \rangle $ is given by:

$$
|\mathbf{v}| = \sqrt{v_1^2 + v_2^2}
$$

In 3D, for the vector $ \mathbf{v} = \langle v_1, v_2, v_3 \rangle $, the magnitude is:

$$
|\mathbf{v}| = \sqrt{v_1^2 + v_2^2 + v_3^2}
$$

### Applications of Vectors

Vectors are used in various applications, such as:
- Describing forces in physics.
- Representing directions and positions in geometry and computer graphics.
- Describing velocity, acceleration, and other physical quantities in kinematics.

### Summary

Vectors are powerful mathematical tools that enable us to model and solve problems involving magnitude and direction. They are fundamental in many fields, including physics, engineering, and computer science, and are crucial for understanding concepts like motion, forces, and geometry.

## Normalizing a Vector

To make a vector's length equal to 1, we need to **normalize** it. This is done by multiplying the vector by a scalar such that the resulting vector has a magnitude (length) of 1. The scalar used to normalize the vector is the **reciprocal of the vector's magnitude**.

### Steps to Normalize a Vector

1. **Find the magnitude of the vector**:
   The magnitude (length) of a vector $ \mathbf{a} = \langle a_1, a_2 \rangle $ is given by the formula:

   $$
   |\mathbf{a}| = \sqrt{a_1^2 + a_2^2}
   $$

2. **Divide the vector by its magnitude**:
   To make the vector unit length (magnitude = 1), divide each component of the vector by its magnitude:

   $$
   \mathbf{a}_{\text{normalized}} = \frac{1}{|\mathbf{a}|} \cdot \mathbf{a}
   $$

### Example

Given the vector $ \mathbf{a} = \langle 3, 4 \rangle $, we first calculate its magnitude:

$$
|\mathbf{a}| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5
$$

To make the vector's length equal to 1, we multiply the vector by the reciprocal of its magnitude, which is $ \frac{1}{5} $:

$$
\mathbf{a}_{\text{normalized}} = \frac{1}{5} \cdot \langle 3, 4 \rangle = \langle \frac{3}{5}, \frac{4}{5} \rangle
$$

Thus, the vector $ \mathbf{a} = \langle 3, 4 \rangle $ must be multiplied by $ \frac{1}{5} $ to make its length equal to 1.

### General Formula

For any vector $ \mathbf{a} = \langle a_1, a_2 \rangle $, the scalar to multiply the vector by to make its length equal to 1 is:

$$
\frac{1}{|\mathbf{a}|}
$$

## Length and Unit Vector of a Vector

### Length (Magnitude) of a Vector

The length (or magnitude) of a vector $ \mathbf{b} = \langle b_1, b_2 \rangle $ is given by the formula:

$$
|\mathbf{b}| = \sqrt{b_1^2 + b_2^2}
$$

### Unit Vector

The unit vector in the direction of a vector is obtained by dividing the vector by its magnitude. For a vector $ \mathbf{b} = \langle b_1, b_2 \rangle $, the unit vector $ \hat{\mathbf{b}} $ is:

$$
\hat{\mathbf{b}} = \frac{1}{|\mathbf{b}|} \cdot \mathbf{b}
$$

### Example

Given the vector $ \mathbf{b} = \langle 1, 1 \rangle $, we first calculate its magnitude:

$$
|\mathbf{b}| = \sqrt{1^2 + 1^2} = \sqrt{1 + 1} = \sqrt{2}
$$

Now, to find the unit vector, we divide each component of the vector by its magnitude:

$$
\hat{\mathbf{b}} = \frac{1}{|\mathbf{b}|} \cdot \mathbf{b} = \frac{1}{\sqrt{2}} \cdot \langle 1, 1 \rangle = \langle \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \rangle
$$

Thus, the length of the vector is $ \sqrt{2} $, and the unit vector is:

$$
\hat{\mathbf{b}} = \langle \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \rangle
$$

### Length of a Vector

The **length** (or **magnitude**) of a vector $ \mathbf{c} = [c_1, c_2, c_3] $ in 3-dimensional space is given by the formula:

$
|\mathbf{c}| = \sqrt{c_1^2 + c_2^2 + c_3^2}
$

For the vector $ \mathbf{c} = [1, 2, 3] $, we calculate the length as follows:

$
|\mathbf{c}| = \sqrt{1^2 + 2^2 + 3^2} = \sqrt{1 + 4 + 9} = \sqrt{14}
$

Thus, the length of the vector $ \mathbf{c} = [1, 2, 3] $ is $ \sqrt{14} $.

---

### Unit Vector

A **unit vector** is a vector that has a length of 1. To find the unit vector $ \hat{c} $ of $ \mathbf{c} $, we divide each component of $ \mathbf{c} $ by its length:

$
\hat{c} = \frac{\mathbf{c}}{|\mathbf{c}|}
$

So, the unit vector of $ \mathbf{c} = [1, 2, 3] $ is:

$
\hat{c} = \left[ \frac{1}{\sqrt{14}}, \frac{2}{\sqrt{14}}, \frac{3}{\sqrt{14}} \right]
$

Thus, the unit vector of $ \mathbf{c} $ is:

$
\hat{c} = \left[ \frac{1}{\sqrt{14}}, \frac{2}{\sqrt{14}}, \frac{3}{\sqrt{14}} \right]
$

### Cartesian Coordinates in a New Basis

Given the vector $ \mathbf{v} = [2, 3, 4] $ and the basis vectors $ \mathbf{b_1} = [1, 0, 1] $, $ \mathbf{b_2} = [0, 1, 0] $, and $ \mathbf{b_3} = [1, 0, -1] $, we want to find the coordinates of $ \mathbf{v} $ in the new basis.

To express $ \mathbf{v} $ as a linear combination of the basis vectors $ \mathbf{b_1} $, $ \mathbf{b_2} $, and $ \mathbf{b_3} $, we solve for scalars $ c_1 $, $ c_2 $, and $ c_3 $ such that:

$
\mathbf{v} = c_1 \mathbf{b_1} + c_2 \mathbf{b_2} + c_3 \mathbf{b_3}
$

This gives the system of equations:

$
\begin{bmatrix} 2 \\ 3 \\ 4 \end{bmatrix} = c_1 \begin{bmatrix} 1 \\ 0 \\ 1 \end{bmatrix} + c_2 \begin{bmatrix} 0 \\ 1 \\ 0 \end{bmatrix} + c_3 \begin{bmatrix} 1 \\ 0 \\ -1 \end{bmatrix}
$

This expands to:

$
\begin{bmatrix} 2 \\ 3 \\ 4 \end{bmatrix} = \begin{bmatrix} c_1 + c_3 \\ c_2 \\ c_1 - c_3 \end{bmatrix}
$

Equating the components:

1. $ c_1 + c_3 = 2 $
2. $ c_2 = 3 $
3. $ c_1 - c_3 = 4 $

---

### Solving the System of Equations

From equation 2, we know that $ c_2 = 3 $.

Now, solving the system of equations:

1. $ c_1 + c_3 = 2 $
2. $ c_1 - c_3 = 4 $

Add these two equations:

$
(c_1 + c_3) + (c_1 - c_3) = 2 + 4
$

$
2c_1 = 6 \implies c_1 = 3
$

Now substitute $ c_1 = 3 $ into either equation, for example, the first one:

$
3 + c_3 = 2 \implies c_3 = -1
$

---

### Final Answer

Thus, the coordinates of $ \mathbf{v} $ in the new basis are:

$
c_1 = 3, \, c_2 = 3, \, c_3 = -1
$

So, the Cartesian coordinates of $ \mathbf{v} $ in the basis $ \{ \mathbf{b_1}, \mathbf{b_2}, \mathbf{b_3} \} $ are:

$
[3, 3, -1]
$
