## Limits of Real Functions

### Definition of Limit:
A limit is the value that a function approaches as the input approaches a certain value.

For a function $f(x)$, the limit as $x$ approaches a value $a$ is denoted as:

$
\lim_{x \to a} f(x) = L
$

This means that as $x$ gets closer and closer to $a$, $f(x)$ approaches the value $L$.

### Types of Limits:

1. **Finite Limits**: If a function approaches a finite value as $x$ approaches $a$.
   - Example: $ \lim_{x \to 2} (3x + 1) = 7 $

2. **Infinite Limits**: If a function approaches infinity or negative infinity as $x$ approaches $a$.
   - Example: $ \lim_{x \to 0^+} \frac{1}{x} = \infty $

3. **One-Sided Limits**:
   - **Left-hand limit**: $ \lim_{x \to a^-} f(x) $
   - **Right-hand limit**: $ \lim_{x \to a^+} f(x) $

4. **Limits at Infinity**: The behavior of a function as $x$ approaches infinity ($x \to \infty$) or negative infinity ($x \to -\infty$).
   - Example: $ \lim_{x \to \infty} \frac{1}{x} = 0 $

### Important Limit Properties:

1. **Sum Rule**:
   $
   \lim_{x \to a} [f(x) + g(x)] = \lim_{x \to a} f(x) + \lim_{x \to a} g(x)
   $

2. **Product Rule**:
   $
   \lim_{x \to a} [f(x) \cdot g(x)] = \lim_{x \to a} f(x) \cdot \lim_{x \to a} g(x)
   $

3. **Quotient Rule**:
   $
   \lim_{x \to a} \frac{f(x)}{g(x)} = \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)} \quad \text{if} \quad \lim_{x \to a} g(x) \neq 0
   $

4. **Constant Multiple Rule**:
   $
   \lim_{x \to a} [c \cdot f(x)] = c \cdot \lim_{x \to a} f(x)
   $

5. **Power Rule**:
   $
   \lim_{x \to a} [f(x)]^n = [\lim_{x \to a} f(x)]^n
   $

### L'Hopital's Rule:
When evaluating limits that result in indeterminate forms like $ \frac{0}{0} $ or $ \frac{\infty}{\infty} $, we can apply L'Hopital's Rule:

$
\lim_{x \to a} \frac{f(x)}{g(x)} = \lim_{x \to a} \frac{f'(x)}{g'(x)}
$

provided that the limit on the right-hand side exists.

### Examples:

1. $ \lim_{x \to 3} \frac{x^2 - 9}{x - 3} $
2. $ \lim_{x \to 0} \frac{\sin(x)}{x} $
3. $ \lim_{x \to \infty} \frac{1}{x^2} $

### Indeterminate Forms:
1. $ \frac{0}{0} $
2. $ \frac{\infty}{\infty} $
3. $ 0 \times \infty $
4. $ \infty - \infty $
5. $ 0^0 $, $ \infty^0 $, $ 1^\infty $

### Conclusion:
Limits are an essential concept in calculus and are used to understand the behavior of functions as the input approaches certain values. They are foundational for further topics such as continuity, derivatives, and integrals.


#Example-1

$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}
$



$
= \frac{x^2(x + 2)}{x^3(x - 3)}
$



$
= \frac{x(x + 2)}{x^3(x - 3)}
$



$
= \frac{\frac{x(x + 2)}{x^3}}{\frac{x^3(x - 3)}{x^3}} = \frac{\frac{x + 2}{x^2}}{x - 3}
$



$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} = 0
$

$

$



$
\lim_{x \to 0} \frac{\sin(3x)}{2x + 1} = 0
$
