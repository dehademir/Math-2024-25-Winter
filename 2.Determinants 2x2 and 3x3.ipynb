# Determinants of 2x2 and 3x3 Matrices

The determinant is a scalar value that can be computed from the elements of a square matrix. It provides important properties of the matrix, such as whether it is invertible.

---

## Determinant of a $2 \times 2$ Matrix
For a $2 \times 2$ matrix:
$$A = \begin{bmatrix} a & b \\ c & d \end{bmatrix},$$  
the determinant is calculated as:
$$\text{det}(A) = ad - bc.$$

### Example:
If $A = \begin{bmatrix} 2 & 3 \\ 4 & 5 \end{bmatrix},$  
then:
$$\text{det}(A) = (2)(5) - (3)(4) = 10 - 12 = -2.$$

---

## Determinant of a $3 \times 3$ Matrix
For a $3 \times 3$ matrix:
$$A = \begin{bmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{bmatrix},$$  
the determinant is calculated using cofactor expansion along the first row:
$$\text{det}(A) = a(ei - fh) - b(di - fg) + c(dh - eg).$$

Alternatively, you can expand along any row or column.

### Example:
If $A = \begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{bmatrix},$  
then:
$$\text{det}(A) = 1(5 \cdot 9 - 6 \cdot 8) - 2(4 \cdot 9 - 6 \cdot 7) + 3(4 \cdot 8 - 5 \cdot 7),$$  
$$\text{det}(A) = 1(-3) - 2(-6) + 3(-3),$$  
$$\text{det}(A) = -3 + 12 - 9 = 0.$$

---

## Summary of Steps:
### For $2 \times 2$:
1. Multiply the diagonal elements: $a \cdot d$.
2. Subtract the product of the off-diagonal elements: $b \cdot c$.
3. Result: $\text{det}(A) = ad - bc$.

### For $3 \times 3$:
1. Choose a row or column to expand (commonly the first row).
2. Multiply each element by the determinant of the $2 \times 2$ submatrix formed by removing the element's row and column.
3. Alternate the signs for each term: $+, -, +$ (or $-, +, -$ if expanding along a column).

---

## Properties of Determinants:
1. A matrix is invertible if and only if its determinant is non-zero.
2. The determinant of a triangular matrix (upper or lower) is the product of its diagonal entries.
3. Swapping two rows or columns of a matrix multiplies the determinant by $-1$.
4. If a row or column of a matrix is multiplied by a scalar, the determinant is multiplied by the same scalar.
5. The determinant is $0$ if two rows or columns are identical or if a row/column is all zeros.

# Determinants for $2 \times 2$ Matrices

## Formula
The determinant of a $2 \times 2$ matrix:
$$A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$$  
is given by:
$$\text{det}(A) = ad - bc.$$

---

## Given Matrices

### Matrix $A$:
$$A = \begin{bmatrix} 2 & 3 \\ 1 & 4 \end{bmatrix}$$  
$$\text{det}(A) = (2)(4) - (3)(1) = 8 - 3 = 5.$$

---

### Matrix $B$:
$$B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}$$  
$$\text{det}(B) = (5)(8) - (6)(7) = 40 - 42 = -2.$$

---

### Matrix $C$:
$$C = \begin{bmatrix} -1 & 2 \\ 3 & 0 \end{bmatrix}$$  
$$\text{det}(C) = (-1)(0) - (2)(3) = 0 - 6 = -6.$$

---

## Results
1. $\text{det}(A) = 5$
2. $\text{det}(B) = -2$
3. $\text{det}(C) = -6$

# Determinants for $3 \times 3$ Matrices

## Formula
The determinant of a $3 \times 3$ matrix:  
$$A = \begin{bmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{bmatrix}$$  
is given by the cofactor expansion along the first row:
$$\text{det}(A) = a(ei - fh) - b(di - fg) + c(dh - eg).$$

---

## Given Matrices

### Matrix $D$:
$$D = \begin{bmatrix}
1 & 0 & 2 \\
-1 & 3 & 1 \\
2 & 4 & -2
\end{bmatrix}$$  
Expand along the first row:
$$\text{det}(D) = 1 \cdot (3 \cdot -2 - 1 \cdot 4) - 0 \cdot (-1 \cdot -2 - 1 \cdot 2) + 2 \cdot (-1 \cdot 4 - 3 \cdot 2).$$  
$$\text{det}(D) = 1(-6 - 4) + 2(-4 - 6) = -10 - 20 = -30.$$

---

### Matrix $E$:
$$E = \begin{bmatrix}
3 & 1 & -1 \\
0 & 2 & 4 \\
5 & 3 & 2
\end{bmatrix}$$  
Expand along the first row:
$$\text{det}(E) = 3 \cdot (2 \cdot 2 - 4 \cdot 3) - 1 \cdot (0 \cdot 2 - 4 \cdot 5) + (-1) \cdot (0 \cdot 3 - 2 \cdot 5).$$  
$$\text{det}(E) = 3(4 - 12) - 1(0 - 20) - 1(0 - 10).$$  
$$\text{det}(E) = 3(-8) + 20 + 10 = -24 + 30 = 6.$$

---

### Matrix $F$:
$$F = \begin{bmatrix}
2 & -3 & 1 \\
1 & 4 & -2 \\
1 & 5 & 3
\end{bmatrix}$$  
Expand along the first row:
$$\text{det}(F) = 2 \cdot (4 \cdot 3 - (-2) \cdot 5) - (-3) \cdot (1 \cdot 3 - (-2) \cdot 1) + 1 \cdot (1 \cdot 5 - 4 \cdot 1).$$  
$$\text{det}(F) = 2(12 + 10) + 3(3 + 2) + 1(5 - 4).$$  
$$\text{det}(F) = 2(22) + 3(5) + 1(1) = 44 + 15 + 1 = 60.$$

---

## Results
1. $\text{det}(D) = -30$
2. $\text{det}(E) = 6$
3. $\text{det}(F) = 60$
