# Linear Equations by Gauss Elimination

Gauss Elimination is a method used to solve systems of linear equations by transforming the system into an upper triangular form using row operations. The system is then solved by back substitution.

#### Steps:

1. **Write the augmented matrix** for the system of equations:
   - Convert the system of equations into an augmented matrix with the coefficients of variables on the left and constants on the right.

2. **Forward Elimination**:
   - Perform row operations to eliminate the variables below the pivot elements (starting from the top-left corner).
   - Use row swaps, scaling rows, and adding/subtracting rows to create zeros below the pivot element in each column.

3. **Back Substitution**:
   - Once the matrix is in upper triangular form (all elements below the diagonal are zeros), solve the system starting from the bottom row and substitute values back into the equations above.

#### Example:

Given the system:
$$
x + 2y + 3z = 9
$$
$$
2x + 3y + 4z = 14
$$
$$
3x + 4y + 5z = 20
$$

**Step 1: Augmented Matrix**:
$$
\begin{pmatrix}
1 & 2 & 3 & | & 9 \\
2 & 3 & 4 & | & 14 \\
3 & 4 & 5 & | & 20
\end{pmatrix}
$$

**Step 2: Forward Elimination**:
- Row 2 = Row 2 - 2 * Row 1
- Row 3 = Row 3 - 3 * Row 1

After these operations, the matrix becomes:
$$
\begin{pmatrix}
1 & 2 & 3 & | & 9 \\
0 & -1 & -2 & | & -4 \\
0 & -2 & -4 & | & -7
\end{pmatrix}
$$

- Row 3 = Row 3 - 2 * Row 2

Now the matrix becomes:
$$
\begin{pmatrix}
1 & 2 & 3 & | & 9 \\
0 & -1 & -2 & | & -4 \\
0 & 0 & 0 & | & 1
\end{pmatrix}
$$

**Step 3: Back Substitution**:
- The third row gives us the equation $0 = 1$, which is inconsistent.

#### Conclusion:
Since we obtain an inconsistent equation during back substitution, the system has **no solution**.

### Solving Systems of Linear Equations using Gauss Elimination

#### System 1:
$$
x + 2y - 2z = 4
$$
$$
2x + y + z = 0
$$
$$
3x + 2y + z = 1
$$

**Step 1: Augmented Matrix**
$$
\begin{pmatrix}
1 & 2 & -2 & | & 4 \\
2 & 1 & 1 & | & 0 \\
3 & 2 & 1 & | & 1
\end{pmatrix}
$$

**Step 2: Forward Elimination**
- Row 2 = Row 2 - 2 * Row 1
- Row 3 = Row 3 - 3 * Row 1

After performing these operations:
$$
\begin{pmatrix}
1 & 2 & -2 & | & 4 \\
0 & -3 & 5 & | & -8 \\
0 & -4 & 7 & | & -11
\end{pmatrix}
$$

- Row 3 = Row 3 - (4/3) * Row 2

Now the matrix becomes:
$$
\begin{pmatrix}
1 & 2 & -2 & | & 4 \\
0 & -3 & 5 & | & -8 \\
0 & 0 & 0 & | & 0
\end{pmatrix}
$$

This system has **infinitely many solutions** because the third row represents the equation $0 = 0$.

#### System 2:
$$
x + y + z - t = 2
$$
$$
2x + y + z = 3
$$
$$
-x + z - t = 0
$$
$$
3x + 2y - z + 2t = -1
$$

**Step 1: Augmented Matrix**
$$
\begin{pmatrix}
1 & 1 & 1 & -1 & | & 2 \\
2 & 1 & 1 & 0 & | & 3 \\
-1 & 0 & 1 & -1 & | & 0 \\
3 & 2 & -1 & 2 & | & -1
\end{pmatrix}
$$

**Step 2: Forward Elimination**
- Row 2 = Row 2 - 2 * Row 1
- Row 3 = Row 3 + Row 1
- Row 4 = Row 4 - 3 * Row 1

After performing these operations:
$$
\begin{pmatrix}
1 & 1 & 1 & -1 & | & 2 \\
0 & -1 & -1 & 2 & | & -1 \\
0 & 1 & 2 & -2 & | & 2 \\
0 & -1 & -4 & 5 & | & -7
\end{pmatrix}
$$

- Row 3 = Row 3 + Row 2
- Row 4 = Row 4 + Row 2

Now the matrix becomes:
$$
\begin{pmatrix}
1 & 1 & 1 & -1 & | & 2 \\
0 & -1 & -1 & 2 & | & -1 \\
0 & 0 & 1 & 0 & | & 1 \\
0 & 0 & -5 & 7 & | & -8
\end{pmatrix}
$$

- Row 4 = Row 4 + 5 * Row 3

Now the matrix becomes:
$$
\begin{pmatrix}
1 & 1 & 1 & -1 & | & 2 \\
0 & -1 & -1 & 2 & | & -1 \\
0 & 0 & 1 & 0 & | & 1 \\
0 & 0 & 0 & 7 & | & -3
\end{pmatrix}
$$

**Step 3: Back Substitution**
- From the fourth row, we get $7t = -3$, so $t = -\frac{3}{7}$.
- From the third row, $z = 1$.
- From the second row, $-y - z + 2t = -1$ gives $y = -\frac{1}{7}$.
- From the first row, $x + y + z - t = 2$ gives $x = \frac{27}{7}$.

Thus, the solution is:
$$
x = \frac{27}{7}, \quad y = -\frac{1}{7}, \quad z = 1, \quad t = -\frac{3}{7}
$$

#### System 3:
$$
x + y - z - t = 0
$$
$$
2x + 3y - 2z + t = 4
$$
$$
3x + 5z = 0
$$
$$
-x + y - 3z + 2t = 3
$$

**Step 1: Augmented Matrix**
$$
\begin{pmatrix}
1 & 1 & -1 & -1 & | & 0 \\
2 & 3 & -2 & 1 & | & 4 \\
3 & 0 & 5 & 0 & | & 0 \\
-1 & 1 & -3 & 2 & | & 3
\end{pmatrix}
$$

**Step 2: Forward Elimination**
- Row 2 = Row 2 - 2 * Row 1
- Row 3 = Row 3 - 3 * Row 1
- Row 4 = Row 4 + Row 1

After performing these operations:
$$
\begin{pmatrix}
1 & 1 & -1 & -1 & | & 0 \\
0 & 1 & 0 & 3 & | & 4 \\
0 & -3 & 8 & 3 & | & 0 \\
0 & 2 & -4 & 1 & | & 3
\end{pmatrix}
$$

- Row 3 = Row 3 + 3 * Row 2
- Row 4 = Row 4 - 2 * Row 2

Now the matrix becomes:
$$
\begin{pmatrix}
1 & 1 & -1 & -1 & | & 0 \\
0 & 1 & 0 & 3 & | & 4 \\
0 & 0 & 8 & 12 & | & 12 \\
0 & 0 & -4 & -5 & | & -5
\end{pmatrix}
$$

- Row 4 = Row 4 + \(\frac{1}{2}\) * Row 3

Now the matrix becomes:
$$
\begin{pmatrix}
1 & 1 & -1 & -1 & | & 0 \\
0 & 1 & 0 & 3 & | & 4 \\
0 & 0 & 8 & 12 & | & 12 \\
0 & 0 & 0 & 0 & | & 0
\end{pmatrix}
$$

This system has **infinitely many solutions** because the last row represents the equation $0 = 0$.
