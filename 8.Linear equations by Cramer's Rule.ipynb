# Linear Equations by Cramer's Rule

Cramer's Rule provides a method for solving systems of linear equations using determinants. It is particularly useful for systems of equations with as many equations as unknowns. Cramer's Rule can be applied to a system of linear equations represented in matrix form as:

$$
A \mathbf{x} = \mathbf{b}
$$

Where:
- $A$ is a square matrix representing the coefficients of the system.
- $\mathbf{x}$ is the column vector of variables.
- $\mathbf{b}$ is the column vector representing the constants.

#### General Form of Cramer's Rule:
For a system of linear equations:
$$
a_1x_1 + a_2x_2 + \dots + a_nx_n = b_1
$$
$$
a_1x_1 + a_2x_2 + \dots + a_nx_n = b_2
$$
$$
\vdots
$$
$$
a_1x_1 + a_2x_2 + \dots + a_nx_n = b_n
$$

The solution for each variable $x_i$ is given by:
$$
x_i = \frac{\text{det}(A_i)}{\text{det}(A)}
$$

Where:
- $\text{det}(A)$ is the determinant of the coefficient matrix $A$.
- $A_i$ is the matrix formed by replacing the $i$-th column of $A$ with the column vector $\mathbf{b}$.

#### Steps to Apply Cramer's Rule:
1. **Write the system of equations** in matrix form: $A \mathbf{x} = \mathbf{b}$.
   
2. **Find the determinant of the coefficient matrix** $A$, denoted as $\text{det}(A)$. If $\text{det}(A) = 0$, the system has no unique solution.

3. **Construct the matrices $A_i$** by replacing the $i$-th column of $A$ with $\mathbf{b}$ for each variable $x_i$.

4. **Calculate the determinants of each matrix** $A_i$, denoted $\text{det}(A_i)$.

5. **Solve for each variable** $x_i$ using:
   $$
   x_i = \frac{\text{det}(A_i)}{\text{det}(A)}
   $$

#### Example 1: Solving a 2x2 system using Cramer's Rule
Consider the system:
$$
2x + 3y = 5
$$
$$
4x - y = 3
$$

Matrix form:
$$
A = \begin{pmatrix} 2 & 3 \\ 4 & -1 \end{pmatrix}, \quad \mathbf{x} = \begin{pmatrix} x \\ y \end{pmatrix}, \quad \mathbf{b} = \begin{pmatrix} 5 \\ 3 \end{pmatrix}
$$

##### Step 1: Find the determinant of $A$:
$$
\text{det}(A) = (2)(-1) - (3)(4) = -2 - 12 = -14
$$

##### Step 2: Construct $A_1$ and $A_2$:
- Replace the first column of $A$ with $\mathbf{b}$ for $A_1$:
  $$
  A_1 = \begin{pmatrix} 5 & 3 \\ 3 & -1 \end{pmatrix}
  $$

- Replace the second column of $A$ with $\mathbf{b}$ for $A_2$:
  $$
  A_2 = \begin{pmatrix} 2 & 5 \\ 4 & 3 \end{pmatrix}
  $$

##### Step 3: Find the determinants of $A_1$ and $A_2$:
$$
\text{det}(A_1) = (5)(-1) - (3)(3) = -5 - 9 = -14
$$
$$
\text{det}(A_2) = (2)(3) - (5)(4) = 6 - 20 = -14
$$

##### Step 4: Solve for $x$ and $y$:
$$
x = \frac{\text{det}(A_1)}{\text{det}(A)} = \frac{-14}{-14} = 1
$$
$$
y = \frac{\text{det}(A_2)}{\text{det}(A)} = \frac{-14}{-14} = 1
$$

##### **Solution:**
$$
x = 1, \quad y = 1
$$

---

#### Example 2: Solving a 3x3 system using Cramer's Rule
Consider the system:
$$
x + 2y + 3z = 9
$$
$$
4x + 5y + 6z = 24
$$
$$
7x + 8y + 9z = 39
$$

Matrix form:
$$
A = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{pmatrix}, \quad \mathbf{x} = \begin{pmatrix} x \\ y \\ z \end{pmatrix}, \quad \mathbf{b} = \begin{pmatrix} 9 \\ 24 \\ 39 \end{pmatrix}
$$

##### Step 1: Find the determinant of $A$:
$$
\text{det}(A) = 1(5 \times 9 - 6 \times 8) - 2(4 \times 9 - 6 \times 7) + 3(4 \times 8 - 5 \times 7) = 1(-3) - 2(6) + 3(7) = -3 - 12 + 21 = 6
$$

##### Step 2: Construct $A_1$, $A_2$, and $A_3$:
- Replace the first column of $A$ with $\mathbf{b}$ for $A_1$:
  $$
  A_1 = \begin{pmatrix} 9 & 2 & 3 \\ 24 & 5 & 6 \\ 39 & 8 & 9 \end{pmatrix}
  $$

- Replace the second column of $A$ with $\mathbf{b}$ for $A_2$:
  $$
  A_2 = \begin{pmatrix} 1 & 9 & 3 \\ 4 & 24 & 6 \\ 7 & 39 & 9 \end{pmatrix}
  $$

- Replace the third column of $A$ with $\mathbf{b}$ for $A_3$:
  $$
  A_3 = \begin{pmatrix} 1 & 2 & 9 \\ 4 & 5 & 24 \\ 7 & 8 & 39 \end{pmatrix}
  $$

##### Step 3: Find the determinants of $A_1$, $A_2$, and $A_3$:
$$
\text{det}(A_1) = 6, \quad \text{det}(A_2) = 6, \quad \text{det}(A_3) = 6
$$

##### Step 4: Solve for $x$, $y$, and $z$:
$$
x = \frac{\text{det}(A_1)}{\text{det}(A)} = \frac{6}{6} = 1
$$
$$
y = \frac{\text{det}(A_2)}{\text{det}(A)} = \frac{6}{6} = 1
$$
$$
z = \frac{\text{det}(A_3)}{\text{det}(A)} = \frac{6}{6} = 1
$$

##### **Solution:**
$$
x = 1, \quad y = 1, \quad z = 1
$$

---

### Important Notes:
- **Determinant must be non-zero**: Cramer's Rule is applicable only when the determinant of the coefficient matrix $A$ is non-zero.
- **Determinants must be computed carefully**: Ensure correct calculation of determinants for each matrix $A$, $A_1$, $A_2$, etc.
- **Efficient for small systems**: Cramer's Rule is most useful for systems with small numbers of variables (e.g., 2x2 or 3x3 systems).



#Example-1

### Solving the System of Equations Using Cramer's Rule

Consider the system of equations:
$$
2x_1 - 3x_2 = 7
$$
$$
3x_1 + 5x_2 = 2
$$

We can solve this system using **Cramer's Rule**.

#### Step 1: Write the system in matrix form:
The system of equations can be written as:
$$
A \mathbf{x} = \mathbf{b}
$$
Where:
$$
A = \begin{pmatrix} 2 & -3 \\ 3 & 5 \end{pmatrix}, \quad \mathbf{x} = \begin{pmatrix} x_1 \\ x_2 \end{pmatrix}, \quad \mathbf{b} = \begin{pmatrix} 7 \\ 2 \end{pmatrix}
$$

#### Step 2: Find the determinant of matrix $A$:
$$
\text{det}(A) = (2)(5) - (-3)(3) = 10 + 9 = 19
$$

#### Step 3: Construct $A_1$ and $A_2$:
- For $x_1$, replace the first column of $A$ with $\mathbf{b}$:
  $$
  A_1 = \begin{pmatrix} 7 & -3 \\ 2 & 5 \end{pmatrix}
  $$

- For $x_2$, replace the second column of $A$ with $\mathbf{b}$:
  $$
  A_2 = \begin{pmatrix} 2 & 7 \\ 3 & 2 \end{pmatrix}
  $$

#### Step 4: Find the determinants of $A_1$ and $A_2$:
$$
\text{det}(A_1) = (7)(5) - (-3)(2) = 35 + 6 = 41
$$
$$
\text{det}(A_2) = (2)(2) - (7)(3) = 4 - 21 = -17
$$

#### Step 5: Solve for $x_1$ and $x_2$:
Using Cramer's Rule:
$$
x_1 = \frac{\text{det}(A_1)}{\text{det}(A)} = \frac{41}{19} = \frac{41}{19}
$$
$$
x_2 = \frac{\text{det}(A_2)}{\text{det}(A)} = \frac{-17}{19} = \frac{-17}{19}
$$

##### **Solution:**
$$
x_1 = \frac{41}{19}, \quad x_2 = \frac{-17}{19}
$$

### Solving the System Using Cramer's Rule

Given the system:
$$
x + y + z - t = 2
$$
$$
x - z + 2t = 6
$$
$$
2x - 3y + t = 4
$$
$$
3x + y + 3z - 4t = -2
$$

1. **Matrix Form**:
$$
A = \begin{pmatrix} 1 & 1 & 1 & -1 \\ 1 & 0 & -1 & 2 \\ 2 & -3 & 0 & 1 \\ 3 & 1 & 3 & -4 \end{pmatrix}, \quad \mathbf{b} = \begin{pmatrix} 2 \\ 6 \\ 4 \\ -2 \end{pmatrix}
$$

2. **Determinant of A**:
$$
\text{det}(A) = -20
$$

3. **Matrices for Cramer's Rule**:
- $A_1$: Replace first column with $\mathbf{b}$:
  $$ \text{det}(A_1) = -40 $$
- $A_2$: Replace second column with $\mathbf{b}$:
  $$ \text{det}(A_2) = -60 $$
- $A_3$: Replace third column with $\mathbf{b}$:
  $$ \text{det}(A_3) = -40 $$
- $A_4$: Replace fourth column with $\mathbf{b}$:
  $$ \text{det}(A_4) = -80 $$

4. **Solution**:
$$
x = \frac{\text{det}(A_1)}{\text{det}(A)} = \frac{-40}{-20} = 2, \quad y = \frac{\text{det}(A_2)}{\text{det}(A)} = \frac{-60}{-20} = 3, \quad z = \frac{\text{det}(A_3)}{\text{det}(A)} = \frac{-40}{-20} = 2, \quad t = \frac{\text{det}(A_4)}{\text{det}(A)} = \frac{-80}{-20} = 4
$$

### Why Cramer's Rule Can't Be Used

Given the system of equations:
$$
x_1 + 2x_2 + 3x_3 = 3
$$
$$
4x_1 + 5x_2 + 6x_3 = 2
$$
$$
7x_1 + 8x_2 + 9x_3 = 1
$$

1. **Matrix Form**:
$$
A = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{pmatrix}, \quad \mathbf{b} = \begin{pmatrix} 3 \\ 2 \\ 1 \end{pmatrix}
$$

2. **Determinant of A**:
The determinant of matrix $A$ is:
$$
\text{det}(A) = 1 \cdot \text{det}\begin{pmatrix} 5 & 6 \\ 8 & 9 \end{pmatrix} - 2 \cdot \text{det}\begin{pmatrix} 4 & 6 \\ 7 & 9 \end{pmatrix} + 3 \cdot \text{det}\begin{pmatrix} 4 & 5 \\ 7 & 8 \end{pmatrix}
$$
But the determinant turns out to be:
$$
\text{det}(A) = 0
$$

3. **Conclusion**:
Since $\text{det}(A) = 0$, the matrix $A$ is **singular** and has no inverse. Therefore, **Cramer's rule cannot be applied** to this system because the determinant of the coefficient matrix must be non-zero for the rule to be valid.
