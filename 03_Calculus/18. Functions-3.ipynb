```
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(x)

def g(x):
    return np.log(x)

def f_of_g(x):
    return np.exp(np.log(x))

def g_of_f(x):
    return np.log(np.exp(x))

x_values_f = np.linspace(-2, 2, 400)
x_values_g = np.linspace(0.1, 2, 400)

y_values_f = f(x_values_f)
y_values_g = g(x_values_g)
y_values_f_of_g = f_of_g(x_values_g)
y_values_g_of_f = g_of_f(x_values_f)

plt.figure(figsize=(10, 6))

plt.plot(x_values_f, y_values_f, label=r'$f(x) = e^x$', color='blue')
plt.plot(x_values_g, y_values_g, label=r'$g(x) = \ln(x)$', color='green')
plt.plot(x_values_g, y_values_f_of_g, label=r'$f(g(x))$', color='red', linestyle='--')
plt.plot(x_values_f, y_values_g_of_f, label=r'$g(f(x))$', color='purple', linestyle='--')

plt.title('Functions and Their Compositions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.grid(True)

plt.show()
```
