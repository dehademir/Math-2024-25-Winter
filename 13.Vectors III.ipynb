#Example-1

We are given two points $ A(-1, 2) $ and $ B(3, -2) $, and we need to find the point that divides the line segment joining $ A $ and $ B $ in the ratio $ 1:3 $.

### Step 1: Use the section formula

The section formula for dividing a line segment in the ratio $ m:n $ is given by:

$
P = \left( \frac{n \cdot x_1 + m \cdot x_2}{m + n}, \frac{n \cdot y_1 + m \cdot y_2}{m + n} \right)
$

Where:
- $ P(x, y) $ is the point dividing the line segment in the ratio $ m:n $,
- $ A(x_1, y_1) $ and $ B(x_2, y_2) $ are the coordinates of the points.

### Step 2: Apply the section formula

For the given ratio $ 1:3 $, we have:
- $ m = 1 $,
- $ n = 3 $,
- $ A(-1, 2) $ and $ B(3, -2) $.

Substitute into the section formula:

$
P = \left( \frac{3 \cdot (-1) + 1 \cdot 3}{1 + 3}, \frac{3 \cdot 2 + 1 \cdot (-2)}{1 + 3} \right)
$

Simplifying:

$
P = \left( \frac{-3 + 3}{4}, \frac{6 - 2}{4} \right)
$

$
P = \left( \frac{0}{4}, \frac{4}{4} \right)
$

$
P = (0, 1)
$

So, the point dividing the line segment in the ratio $ 1:3 $ is $ P(0, 1) $.

### Step 3: Plot the points and the division

We will plot the points $ A(-1, 2) $, $ B(3, -2) $, and $ P(0, 1) $ on a graph to illustrate the result.


```
import matplotlib.pyplot as plt

# Coordinates of points A, B, and P
A = [-1, 2]
B = [3, -2]
P = [0, 1]

# Create a plot
plt.figure(figsize=(6,6))

# Plot points A, B, and P
plt.plot(A[0], A[1], 'go', label="Point A (-1, 2)")
plt.plot(B[0], B[1], 'ro', label="Point B (3, -2)")
plt.plot(P[0], P[1], 'bo', label="Point P (0, 1)")

# Plot the line segment AB
plt.plot([A[0], B[0]], [A[1], B[1]], 'k-', label="Line segment AB")

# Label the points
plt.text(A[0], A[1], ' A', fontsize=12, ha='right')
plt.text(B[0], B[1], ' B', fontsize=12, ha='right')
plt.text(P[0], P[1], ' P', fontsize=12, ha='right')

# Add grid, labels, and title
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlim(-2, 4)
plt.ylim(-3, 3)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Division of Line Segment AB in Ratio 1:3')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

#Example-2

We are given the vector $ \mathbf{a} = (3, 4) $, and we need to project it onto the $ OX $ and $ OY $ axes.

### Step 1: Project onto the $ OX $ axis

The projection of a vector $ \mathbf{a} = (a_x, a_y) $ onto the $ OX $ axis is given by:

$
\text{Projection on } OX = (a_x, 0)
$

For $ \mathbf{a} = (3, 4) $, the projection onto the $ OX $ axis is:

$
\text{Projection on } OX = (3, 0)
$

### Step 2: Project onto the $ OY $ axis

The projection of a vector $ \mathbf{a} = (a_x, a_y) $ onto the $ OY $ axis is given by:

$
\text{Projection on } OY = (0, a_y)
$

For $ \mathbf{a} = (3, 4) $, the projection onto the $ OY $ axis is:

$
\text{Projection on } OY = (0, 4)
$

### Step 3: Plot the vector and its projections

We will now plot the vector $ \mathbf{a} = (3, 4) $, its projection onto the $ OX $ axis, and its projection onto the $ OY $ axis on a graph.


```
import matplotlib.pyplot as plt
import numpy as np

# Coordinates of the vector a and its projections
a = [3, 4]
projection_OX = [3, 0]
projection_OY = [0, 4]

# Create a plot
plt.figure(figsize=(6,6))

# Plot the vector a
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='b', label="Vector a (3, 4)")

# Plot the projections onto OX and OY axes
plt.quiver(0, 0, projection_OX[0], projection_OX[1], angles='xy', scale_units='xy', scale=1, color='g', label="Projection on OX (3, 0)")
plt.quiver(0, 0, projection_OY[0], projection_OY[1], angles='xy', scale_units='xy', scale=1, color='r', label="Projection on OY (0, 4)")

# Setting the limits of the graph
plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels, title, and grid
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Projection of Vector a onto OX and OY Axes')
plt.grid(True)

# Add legend
plt.legend()

# Show the plot
plt.show()
```

#Example-3

We are given the vectors $ \mathbf{a} = (2, 3) $ and $ \mathbf{b} = (1, 1) $, and we need to project $ \mathbf{a} $ onto $ \mathbf{b} $.

### Step 1: Use the projection formula

The projection of a vector $ \mathbf{a} = (a_x, a_y) $ onto a vector $ \mathbf{b} = (b_x, b_y) $ is given by the formula:

$
\text{proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{\mathbf{b} \cdot \mathbf{b}} \mathbf{b}
$

Where:
- $ \mathbf{a} \cdot \mathbf{b} $ is the dot product of $ \mathbf{a} $ and $ \mathbf{b} $,
- $ \mathbf{b} \cdot \mathbf{b} $ is the dot product of $ \mathbf{b} $ with itself.

### Step 2: Compute the projection

For $ \mathbf{a} = (2, 3) $ and $ \mathbf{b} = (1, 1) $, calculate the dot products:

$
\mathbf{a} \cdot \mathbf{b} = 2 \cdot 1 + 3 \cdot 1 = 5
$
$
\mathbf{b} \cdot \mathbf{b} = 1 \cdot 1 + 1 \cdot 1 = 2
$

Now, compute the projection:

$
\text{proj}_{\mathbf{b}} \mathbf{a} = \frac{5}{2} \cdot (1, 1) = \left( \frac{5}{2}, \frac{5}{2} \right)
$

Thus, the projection of $ \mathbf{a} $ onto $ \mathbf{b} $ is $ \left( \frac{5}{2}, \frac{5}{2} \right) $.

### Step 3: Plot the vectors and the projection

We will now plot the vector $ \mathbf{a} = (2, 3) $, the vector $ \mathbf{b} = (1, 1) $, and the projection of $ \mathbf{a} $ onto $ \mathbf{b} = \left( \frac{5}{2}, \frac{5}{2} \right) $ on a graph.


```
import matplotlib.pyplot as plt
import numpy as np

# Coordinates of the vectors a, b, and their projection
a = [2, 3]
b = [1, 1]
projection = [5/2, 5/2]

# Create a plot
plt.figure(figsize=(6,6))

# Plot the vector a
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='b', label="Vector a (2, 3)")

# Plot the vector b
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='g', label="Vector b (1, 1)")

# Plot the projection of a onto b
plt.quiver(0, 0, projection[0], projection[1], angles='xy', scale_units='xy', scale=1, color='r', label="Projection of a onto b (5/2, 5/2)")

# Setting the limits of the graph
plt.xlim(-1, 3)
plt.ylim(-1, 4)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels, title, and grid
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Projection of Vector a onto Vector b')
plt.grid(True)

# Add legend
plt.legend()

# Show the plot
plt.show()
```

#Example-4

We are given the vectors $ \mathbf{b} = (1, 1) $ and $ \mathbf{a} = (2, 3) $, and we need to project $ \mathbf{b} $ onto $ \mathbf{a} $.

### Step 1: Use the projection formula

The projection of a vector $ \mathbf{b} = (b_x, b_y) $ onto a vector $ \mathbf{a} = (a_x, a_y) $ is given by the formula:

$
\text{proj}_{\mathbf{a}} \mathbf{b} = \frac{\mathbf{b} \cdot \mathbf{a}}{\mathbf{a} \cdot \mathbf{a}} \mathbf{a}
$

Where:
- $ \mathbf{b} \cdot \mathbf{a} $ is the dot product of $ \mathbf{b} $ and $ \mathbf{a} $,
- $ \mathbf{a} \cdot \mathbf{a} $ is the dot product of $ \mathbf{a} $ with itself.

### Step 2: Compute the projection

For $ \mathbf{b} = (1, 1) $ and $ \mathbf{a} = (2, 3) $, calculate the dot products:

$
\mathbf{b} \cdot \mathbf{a} = 1 \cdot 2 + 1 \cdot 3 = 5
$
$
\mathbf{a} \cdot \mathbf{a} = 2 \cdot 2 + 3 \cdot 3 = 13
$

Now, compute the projection:

$
\text{proj}_{\mathbf{a}} \mathbf{b} = \frac{5}{13} \cdot (2, 3) = \left( \frac{10}{13}, \frac{15}{13} \right)
$

Thus, the projection of $ \mathbf{b} $ onto $ \mathbf{a} $ is $ \left( \frac{10}{13}, \frac{15}{13} \right) $.

### Step 3: Plot the vectors and the projection

We will now plot the vector $ \mathbf{b} = (1, 1) $, the vector $ \mathbf{a} = (2, 3) $, and the projection of $ \mathbf{b} $ onto $ \mathbf{a} = \left( \frac{10}{13}, \frac{15}{13} \right) $ on a graph.


```
import matplotlib.pyplot as plt
import numpy as np

# Coordinates of the vectors b, a, and their projection
b = [1, 1]
a = [2, 3]
projection = [10/13, 15/13]

# Create a plot
plt.figure(figsize=(6,6))

# Plot the vector b
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label="Vector b (1, 1)")

# Plot the vector a
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='g', label="Vector a (2, 3)")

# Plot the projection of b onto a
plt.quiver(0, 0, projection[0], projection[1], angles='xy', scale_units='xy', scale=1, color='r', label="Projection of b onto a (10/13, 15/13)")

# Setting the limits of the graph
plt.xlim(-1, 3)
plt.ylim(-1, 4)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels, title, and grid
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Projection of Vector b onto Vector a')
plt.grid(True)

# Add legend
plt.legend()

# Show the plot
plt.show()
```
