# Determinants from the Gauss Method and Triangular Matrices

#### 1. Introduction to the Gauss Method
The **Gauss Method** (or Gaussian Elimination) is a method used to simplify a matrix into a form where the determinant can be easily calculated. The goal is to transform the matrix into an upper triangular form using a series of row operations.

##### Row Operations:
The following row operations can be performed without changing the determinant:
1. **Swapping two rows**: The determinant changes sign.
2. **Multiplying a row by a scalar**: The determinant is multiplied by that scalar.
3. **Adding a multiple of one row to another**: The determinant remains unchanged.

#### 2. Gauss Method Process
To find the determinant using the Gauss method, follow these steps:

1. **Transform the matrix into upper triangular form**:
   Use Gaussian elimination (row reduction) to convert the matrix to an upper triangular matrix (where all entries below the main diagonal are zero).

2. **Calculate the determinant**:
   The determinant of an upper triangular matrix is the product of the diagonal elements.
   $$
   \text{det}(A) = \prod_{i=1}^{n} a_{ii}
   $$
   where $a_{ii}$ are the diagonal elements of the triangular matrix.

3. **Sign adjustment**:
   If any row swaps were performed during the elimination process, adjust the determinant by multiplying it by $(-1)^k$ where $k$ is the number of row swaps.

#### 3. Example using the Gauss Method

Letâ€™s compute the determinant of the matrix:

$$
A = \begin{pmatrix}
2 & 1 & 3 \\
1 & 2 & 1 \\
3 & 1 & 2
\end{pmatrix}
$$

##### Step 1: Transforming to Upper Triangular Form
Perform row operations to create zeros below the diagonal.

1. Subtract $\frac{1}{2}$ of row 1 from row 2:
   $$
   R_2 \rightarrow R_2 - \frac{1}{2} R_1
   $$
   The matrix becomes:
   $$
   A = \begin{pmatrix}
   2 & 1 & 3 \\
   0 & \frac{3}{2} & -\frac{1}{2} \\
   3 & 1 & 2
   \end{pmatrix}
   $$

2. Subtract $\frac{3}{2}$ of row 1 from row 3:
   $$
   R_3 \rightarrow R_3 - \frac{3}{2} R_1
   $$
   The matrix becomes:
   $$
   A = \begin{pmatrix}
   2 & 1 & 3 \\
   0 & \frac{3}{2} & -\frac{1}{2} \\
   0 & -\frac{1}{2} & -\frac{1}{2}
   \end{pmatrix}
   $$

3. Subtract $-\frac{1}{3}$ of row 2 from row 3:
   $$
   R_3 \rightarrow R_3 - \left( -\frac{1}{3} \right) R_2
   $$
   The matrix becomes:
   $$
   A = \begin{pmatrix}
   2 & 1 & 3 \\
   0 & \frac{3}{2} & -\frac{1}{2} \\
   0 & 0 & \frac{1}{6}
   \end{pmatrix}
   $$

##### Step 2: Calculating the Determinant
Now that the matrix is in upper triangular form, we can calculate the determinant by multiplying the diagonal elements:
$$
\text{det}(A) = 2 \times \frac{3}{2} \times \frac{1}{6} = \frac{3}{6} = \frac{1}{2}
$$

Thus, the determinant of matrix $A$ is $\frac{1}{2}$.

#### 4. Triangular Matrices
A **triangular matrix** is a square matrix where all the elements above or below the main diagonal are zero.

- An **upper triangular matrix** has zeros below the main diagonal.
- A **lower triangular matrix** has zeros above the main diagonal.

For any triangular matrix, the determinant is simply the product of its diagonal elements. If $T$ is a triangular matrix, then:
$$
\text{det}(T) = \prod_{i=1}^{n} t_{ii}
$$
where $t_{ii}$ are the diagonal elements of the matrix.

##### Example of Upper Triangular Matrix:

$$
T = \begin{pmatrix}
2 & 1 & 3 \\
0 & 4 & 5 \\
0 & 0 & 6
\end{pmatrix}
$$

The determinant of this upper triangular matrix is:
$$
\text{det}(T) = 2 \times 4 \times 6 = 48
$$

##### Example of Lower Triangular Matrix:

$$
L = \begin{pmatrix}
3 & 0 & 0 \\
1 & 2 & 0 \\
4 & 5 & 6
\end{pmatrix}
$$

The determinant of this lower triangular matrix is:
$$
\text{det}(L) = 3 \times 2 \times 6 = 36
$$

### Conclusion
The Gauss method simplifies a matrix to an upper triangular form, making it easy to calculate the determinant as the product of the diagonal elements. For triangular matrices, the determinant is simply the product of the diagonal elements.

### Row and Column Operations to Reduce Matrices to Upper Triangular Form

#### 1. Matrix A:

Given matrix $A$:
$$
A = \begin{pmatrix}
12 & 3 & -18 \\
-4 & 0 & 0
\end{pmatrix}
$$

##### Step 1: Row Operations to Create Upper Triangular Matrix

1. **Row 2 operation**: Add $\frac{1}{3}$ of row 1 to row 2 to create a zero in the second row, first column:
   $$ R_2 \rightarrow R_2 + \frac{1}{3}R_1 $$

   The matrix becomes:
   $$
   A = \begin{pmatrix}
   12 & 3 & -18 \\
   0 & 1 & -6
   \end{pmatrix}
   $$

##### Step 2: Calculating the Determinant

Now the matrix is in upper triangular form, and the determinant is the product of the diagonal elements:
$$
\text{det}(A) = 12 \times 1 = 12
$$

Thus, the determinant of matrix $A$ is $12$.

#### 2. Matrix B:

Given matrix $B$:
$$
B = \begin{pmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{pmatrix}
$$

##### Step 1: Row Operations to Create Upper Triangular Matrix

1. **Row 2 operation**: Subtract $4$ times row 1 from row 2 to create a zero in the second row, first column:
   $$ R_2 \rightarrow R_2 - 4R_1 $$

   The matrix becomes:
   $$
   B = \begin{pmatrix}
   1 & 2 & 3 \\
   0 & -3 & -6 \\
   7 & 8 & 9
   \end{pmatrix}
   $$

2. **Row 3 operation**: Subtract $7$ times row 1 from row 3 to create a zero in the third row, first column:
   $$ R_3 \rightarrow R_3 - 7R_1 $$

   The matrix becomes:
   $$
   B = \begin{pmatrix}
   1 & 2 & 3 \\
   0 & -3 & -6 \\
   0 & -6 & -12
   \end{pmatrix}
   $$

3. **Row 3 operation**: Add $2$ times row 2 to row 3 to create a zero in the third row, second column:
   $$ R_3 \rightarrow R_3 + 2R_2 $$

   The matrix becomes:
   $$
   B = \begin{pmatrix}
   1 & 2 & 3 \\
   0 & -3 & -6 \\
   0 & 0 & 0
   \end{pmatrix}
   $$

##### Step 2: Calculating the Determinant

Now the matrix is in upper triangular form, and the determinant is the product of the diagonal elements:
$$
\text{det}(B) = 1 \times (-3) \times 0 = 0
$$

Thus, the determinant of matrix $B$ is $0$.

### Conclusion

- The determinant of matrix $A$ is $12$.
- The determinant of matrix $B$ is $0$.
