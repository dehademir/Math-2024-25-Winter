# Solving Linear Equations - Old School Methods

#### 1. **Substitution Method**
The substitution method is used to solve a system of linear equations by solving one equation for one variable and then substituting that expression into the other equation(s).

##### Steps:
1. Solve one equation for one variable.
2. Substitute this expression into the other equation(s).
3. Solve the resulting equation for the other variable.
4. Substitute the found value back into the original equation to find the first variable.

##### Example:
Solve the system of equations:
$$
x + y = 10 \quad \text{(1)}
$$
$$
2x - y = 3 \quad \text{(2)}
$$

**Step 1:** Solve Equation (1) for $y$:
$$
y = 10 - x
$$

**Step 2:** Substitute $y = 10 - x$ into Equation (2):
$$
2x - (10 - x) = 3
$$
Simplify and solve for $x$:
$$
2x - 10 + x = 3
$$
$$
3x = 13 \quad \Rightarrow \quad x = \frac{13}{3}
$$

**Step 3:** Substitute $x = \frac{13}{3}$ into $y = 10 - x$:
$$
y = 10 - \frac{13}{3} = \frac{30}{3} - \frac{13}{3} = \frac{17}{3}
$$

**Solution:** The solution is $x = \frac{13}{3}$ and $y = \frac{17}{3}$.

---

#### 2. **Elimination Method (Addition/Subtraction Method)**

The elimination method eliminates one of the variables by adding or subtracting the equations.

##### Steps:
1. Multiply one or both equations (if necessary) so that the coefficients of one variable are opposites.
2. Add or subtract the equations to eliminate one variable.
3. Solve the resulting equation for the remaining variable.
4. Substitute the value of the found variable into one of the original equations to find the other variable.

##### Example:
Solve the system of equations:
$$
x + 2y = 8 \quad \text{(1)}
$$
$$
2x - y = 3 \quad \text{(2)}
$$

**Step 1:** Multiply Equation (1) by 2 to make the coefficient of $x$ in both equations the same:
$$
2(x + 2y) = 2 \times 8 \quad \Rightarrow \quad 2x + 4y = 16 \quad \text{(3)}
$$

**Step 2:** Subtract Equation (2) from Equation (3) to eliminate $x$:
$$
(2x + 4y) - (2x - y) = 16 - 3
$$
Simplify:
$$
5y = 13 \quad \Rightarrow \quad y = \frac{13}{5}
$$

**Step 3:** Substitute $y = \frac{13}{5}$ into Equation (1) to find $x$:
$$
x + 2\left(\frac{13}{5}\right) = 8
$$
Simplify:
$$
x + \frac{26}{5} = 8 \quad \Rightarrow \quad x = 8 - \frac{26}{5} = \frac{40}{5} - \frac{26}{5} = \frac{14}{5}
$$

**Solution:** The solution is $x = \frac{14}{5}$ and $y = \frac{13}{5}$.

---

#### 3. **Graphical Method**
The graphical method involves plotting each linear equation on a graph and finding the point of intersection.

##### Steps:
1. Rewrite each equation in slope-intercept form: $y = mx + b$.
2. Plot the two lines on a coordinate plane.
3. The point where the lines intersect is the solution to the system of equations.

##### Example:
Solve the system of equations:
$$
x + y = 10 \quad \text{(1)}
$$
$$
2x - y = 3 \quad \text{(2)}
$$

**Step 1:** Rewrite the equations in slope-intercept form:
- Equation (1): $y = 10 - x$
- Equation (2): $y = 2x - 3$

**Step 2:** Plot the two lines on a graph. The intersection point is the solution.

---

### Conclusion:
- **Substitution Method**: Best when one equation is easy to solve for one variable.
- **Elimination Method**: Useful when the coefficients of a variable are easily matched.
- **Graphical Method**: Offers a visual representation of the solution, but is less precise for larger systems.

These methods are foundational for solving systems of linear equations manually.

### Solving Systems of Equations Without Using Matrices

#### 1. **System of Two Variables:**
Solve the following system of equations:
$$
3x - 2y = 5 \quad \text{(1)}
$$
$$
2x + 3y = 7 \quad \text{(2)}
$$

##### Step 1: Solve one equation for one variable
From Equation (1):
$$
3x = 2y + 5 \quad \Rightarrow \quad x = \frac{2y + 5}{3}
$$

##### Step 2: Substitute into the second equation
Substitute $x = \frac{2y + 5}{3}$ into Equation (2):
$$
2\left(\frac{2y + 5}{3}\right) + 3y = 7
$$
Simplify:
$$
\frac{4y + 10}{3} + 3y = 7
$$
Multiply through by 3 to eliminate the fraction:
$$
4y + 10 + 9y = 21
$$
Simplify:
$$
13y + 10 = 21 \quad \Rightarrow \quad 13y = 11 \quad \Rightarrow \quad y = \frac{11}{13}
$$

##### Step 3: Substitute $y$ back into the equation for $x$
Substitute $y = \frac{11}{13}$ into $x = \frac{2y + 5}{3}$:
$$
x = \frac{2\left(\frac{11}{13}\right) + 5}{3} = \frac{\frac{22}{13} + 5}{3} = \frac{\frac{22}{13} + \frac{65}{13}}{3} = \frac{87}{39} = \frac{29}{13}
$$

##### **Solution:**
$$
x = \frac{29}{13}, \quad y = \frac{11}{13}
$$

#### 2. **System of Two Variables:**
Solve the following system:
$$
2x - 3y = 10 \quad \text{(3)}
$$
$$
4x + 5y = 20 \quad \text{(4)}
$$

##### Step 1: Solve one equation for one variable
From Equation (3):
$$
2x = 3y + 10 \quad \Rightarrow \quad x = \frac{3y + 10}{2}
$$

##### Step 2: Substitute into the second equation
Substitute $x = \frac{3y + 10}{2}$ into Equation (4):
$$
4\left(\frac{3y + 10}{2}\right) + 5y = 20
$$
Simplify:
$$
2(3y + 10) + 5y = 20 \quad \Rightarrow \quad 6y + 20 + 5y = 20 \quad \Rightarrow \quad 11y = 0
$$
Solve for $y$:
$$
y = 0
$$

##### Step 3: Substitute $y = 0$ back into the equation for $x$
Substitute $y = 0$ into $x = \frac{3y + 10}{2}$:
$$
x = \frac{3(0) + 10}{2} = \frac{10}{2} = 5
$$

##### **Solution:**
$$
x = 5, \quad y = 0
$$

#### 3. **System of Three Variables:**
Solve the following system:
$$
2x - y + z = 3 \quad \text{(5)}
$$
$$
x + 2y - z = 1 \quad \text{(6)}
$$
$$
3x - y + 2z = 11 \quad \text{(7)}
$$

##### Step 1: Solve one equation for one variable
From Equation (5):
$$
2x - y + z = 3 \quad \Rightarrow \quad z = 3 - 2x + y \quad \text{(8)}
$$

##### Step 2: Substitute into the other two equations
Substitute $z = 3 - 2x + y$ into Equations (6) and (7):

For Equation (6):
$$
x + 2y - (3 - 2x + y) = 1
$$
Simplify:
$$
x + 2y - 3 + 2x - y = 1 \quad \Rightarrow \quad 3x + y = 4 \quad \text{(9)}
$$

For Equation (7):
$$
3x - y + 2(3 - 2x + y) = 11
$$
Simplify:
$$
3x - y + 6 - 4x + 2y = 11 \quad \Rightarrow \quad -x + y = 5 \quad \text{(10)}
$$

##### Step 3: Solve the system of two equations
Solve Equations (9) and (10):
$$
3x + y = 4 \quad \text{(9)}
$$
$$
-x + y = 5 \quad \text{(10)}
$$

From Equation (10), solve for $y$:
$$
y = x + 5 \quad \text{(11)}
$$

Substitute Equation (11) into Equation (9):
$$
3x + (x + 5) = 4
$$
Simplify:
$$
4x + 5 = 4 \quad \Rightarrow \quad 4x = -1 \quad \Rightarrow \quad x = -\frac{1}{4}
$$

##### Step 4: Solve for $y$ and $z$
Substitute $x = -\frac{1}{4}$ into Equation (11):
$$
y = -\frac{1}{4} + 5 = \frac{19}{4}
$$

Substitute $x = -\frac{1}{4}$ and $y = \frac{19}{4}$ into Equation (8):
$$
z = 3 - 2\left(-\frac{1}{4}\right) + \frac{19}{4} = 3 + \frac{1}{2} + \frac{19}{4} = \frac{12}{4} + \frac{1}{2} + \frac{19}{4} = \frac{31}{4}
$$

##### **Solution:**
$$
x = -\frac{1}{4}, \quad y = \frac{19}{4}, \quad z = \frac{31}{4}
$$
