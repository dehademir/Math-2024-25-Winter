# Solving Linear Equations Using Matrix Inversion

Matrix inversion is a method to solve systems of linear equations using the inverse of the coefficient matrix.

#### General Form of System of Linear Equations:

Consider a system of equations in the form:

$$
A \cdot X = B
$$

Where:
- $A$ is the matrix of coefficients (an $n \times n$ matrix).
- $X$ is the column matrix of variables (an $n \times 1$ matrix).
- $B$ is the column matrix of constants (an $n \times 1$ matrix).

To find the solution for $X$, we need to calculate the inverse of matrix $A$, denoted as $A^{-1}$. If $A^{-1}$ exists, the solution is:

$$
X = A^{-1} \cdot B
$$

#### Steps for Solving Using Matrix Inversion:

1. **Write the system as a matrix equation:**
   $$ A \cdot X = B $$

2. **Check if $A$ is invertible:**  
   - Calculate the determinant of $A$ (denoted as $|A|$).
   - If $|A| = 0$, $A$ is **singular**, and the system cannot be solved using matrix inversion.

3. **Find the inverse of matrix $A$:**  
   Use the formula for the inverse of a $2 \times 2$ matrix or apply Gaussian elimination or other matrix inversion methods for larger matrices.

4. **Multiply $A^{-1}$ by $B$ to find $X$:**
   $$ X = A^{-1} \cdot B $$

#### Example: Solving a System of Equations

Consider the system:

$$
x + 2y = 5
$$
$$
3x + 4y = 6
$$

We can write this as:

$$
A = \begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix}, \quad X = \begin{pmatrix} x \\ y \end{pmatrix}, \quad B = \begin{pmatrix} 5 \\ 6 \end{pmatrix}
$$

1. **Find the determinant of $A$:**

$$
|A| = (1)(4) - (2)(3) = 4 - 6 = -2
$$

Since the determinant is non-zero, $A$ is invertible.

2. **Find the inverse of $A$:**

For a $2 \times 2$ matrix:

$$
A^{-1} = \frac{1}{|A|} \begin{pmatrix} d & -b \\ -c & a \end{pmatrix}
$$

For $A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}$:

$$
A^{-1} = \frac{1}{-2} \begin{pmatrix} 4 & -2 \\ -3 & 1 \end{pmatrix} = \begin{pmatrix} -2 & 1 \\ 1.5 & -0.5 \end{pmatrix}
$$

3. **Multiply $A^{-1}$ by $B$:**

$$
X = A^{-1} \cdot B = \begin{pmatrix} -2 & 1 \\ 1.5 & -0.5 \end{pmatrix} \cdot \begin{pmatrix} 5 \\ 6 \end{pmatrix}
$$

Perform the matrix multiplication:

$$
X = \begin{pmatrix} (-2)(5) + (1)(6) \\ (1.5)(5) + (-0.5)(6) \end{pmatrix}
$$

This simplifies to:

$$
X = \begin{pmatrix} -10 + 6 \\ 7.5 - 3 \end{pmatrix} = \begin{pmatrix} -4 \\ 4.5 \end{pmatrix}
$$

Thus, the solution is:

$$
x = -4, \quad y = 4.5
$$

#### Advantages of Using Matrix Inversion:
- It provides a straightforward method for solving systems of linear equations.
- It can be used for any system of equations where the coefficient matrix is invertible.

#### Disadvantages:
- Matrix inversion is computationally expensive for larger systems.
- The method is not applicable if the coefficient matrix is singular (i.e., non-invertible).

### Solving the System of Linear Equations Using the Inverse Matrix Method

We are given the system of equations:

1. $ x + 2y + 3z = 5 $
2. $ 2y + 3z = 4 $
3. $ 3z = 3 $

We aim to solve this system using the inverse matrix method.

#### Step 1: Write the system as a matrix equation.

We can rewrite the system in matrix form as $ A \cdot X = B $, where:

$$
A = \begin{pmatrix}
1 & 2 & 3 \\
0 & 2 & 3 \\
0 & 0 & 3
\end{pmatrix}, \quad X = \begin{pmatrix} x \\ y \\ z \end{pmatrix}, \quad B = \begin{pmatrix} 5 \\ 4 \\ 3 \end{pmatrix}
$$

#### Step 2: Find the determinant of matrix $ A $.

To check if the matrix $ A $ is invertible, we calculate its determinant:

$$
|A| = \begin{vmatrix}
1 & 2 & 3 \\
0 & 2 & 3 \\
0 & 0 & 3
\end{vmatrix}
$$

Since this is a triangular matrix, the determinant is simply the product of the diagonal elements:

$$
|A| = 1 \times 2 \times 3 = 6
$$

Since the determinant is non-zero, the matrix is invertible, and we can proceed.

#### Step 3: Find the inverse of matrix $ A $.

For a triangular matrix, the inverse can be calculated directly. The inverse of $ A $ is:

$$
A^{-1} = \begin{pmatrix}
1 & 0 & 0 \\
0 & \frac{1}{2} & -\frac{1}{2} \\
0 & 0 & \frac{1}{3}
\end{pmatrix}
$$

#### Step 4: Multiply $ A^{-1} $ by $ B $ to find $ X $.

Now, we solve for $ X $ by multiplying $ A^{-1} $ with $ B $:

$$
X = A^{-1} \cdot B = \begin{pmatrix} 1 & 0 & 0 \\ 0 & \frac{1}{2} & -\frac{1}{2} \\ 0 & 0 & \frac{1}{3} \end{pmatrix} \cdot \begin{pmatrix} 5 \\ 4 \\ 3 \end{pmatrix}
$$

Performing the matrix multiplication:

$$
X = \begin{pmatrix}
(1)(5) + (0)(4) + (0)(3) \\
(0)(5) + \frac{1}{2}(4) + \left(-\frac{1}{2}\right)(3) \\
(0)(5) + (0)(4) + \frac{1}{3}(3)
\end{pmatrix}
$$

This simplifies to:

$$
X = \begin{pmatrix}
5 \\
\frac{4}{2} - \frac{3}{2} \\
1
\end{pmatrix}
= \begin{pmatrix}
5 \\
\frac{1}{2} \\
1
\end{pmatrix}
$$

Thus, the solution is:

$$
x = 5, \quad y = \frac{1}{2}, \quad z = 1
$$

### Solving the System of Linear Equations Using the Inverse Matrix Method

We are given the system of equations:

1. $ x_1 + 2x_2 + 3x_3 = 41 $
2. $ 4x_1 + 5x_2 + 6x_3 = 93 $
3. $ 7x_1 + 8x_2 + 9x_3 = 145 $

We aim to solve this system using the inverse matrix method.

#### Step 1: Write the system as a matrix equation.

We can rewrite the system in matrix form as $ A \cdot X = B $, where:

$$
A = \begin{pmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{pmatrix}, \quad X = \begin{pmatrix} x_1 \\ x_2 \\ x_3 \end{pmatrix}, \quad B = \begin{pmatrix} 41 \\ 93 \\ 145 \end{pmatrix}
$$

#### Step 2: Find the determinant of matrix $ A $.

To check if the matrix $ A $ is invertible, we calculate its determinant:

$$
|A| = \begin{vmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{vmatrix}
$$

Using cofactor expansion:

$$
|A| = 1 \cdot \begin{vmatrix} 5 & 6 \\ 8 & 9 \end{vmatrix} - 2 \cdot \begin{vmatrix} 4 & 6 \\ 7 & 9 \end{vmatrix} + 3 \cdot \begin{vmatrix} 4 & 5 \\ 7 & 8 \end{vmatrix}
$$

Calculating the 2x2 determinants:

$$
|A| = 1 \cdot (5 \cdot 9 - 6 \cdot 8) - 2 \cdot (4 \cdot 9 - 6 \cdot 7) + 3 \cdot (4 \cdot 8 - 5 \cdot 7)
$$
$$
|A| = 1 \cdot (45 - 48) - 2 \cdot (36 - 42) + 3 \cdot (32 - 35)
$$
$$
|A| = 1 \cdot (-3) - 2 \cdot (-6) + 3 \cdot (-3)
$$
$$
|A| = -3 + 12 - 9 = 0
$$

Since the determinant of matrix $ A $ is 0, the matrix is **singular**, meaning it is **non-invertible**. Therefore, the system does not have a unique solution.
