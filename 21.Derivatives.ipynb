# Derivatives Notes

## 1. **Definition of Derivative**

The derivative of a function $f(x)$ at a point $x$ is the rate at which $f(x)$ changes with respect to $x$. It is the slope of the tangent line to the curve of the function at that point.

The formal definition of a derivative is:

$
f'(x) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}
$

## 2. **Basic Derivative Rules**

### **Power Rule**
For a function of the form $f(x) = x^n$, the derivative is:

$
f'(x) = n \cdot x^{n-1}
$

Example:
If $f(x) = x^3$, then:

$
f'(x) = 3x^2
$

### **Constant Rule**
The derivative of a constant is always 0.

Example:
If $f(x) = 5$, then:

$
f'(x) = 0
$

### **Sum Rule**
The derivative of the sum of two functions is the sum of their derivatives.

$
(f(x) + g(x))' = f'(x) + g'(x)
$

Example:
If $f(x) = x^2$ and $g(x) = 3x$, then:

$
(f(x) + g(x))' = 2x + 3
$

### **Product Rule**
The derivative of the product of two functions is:

$
(f(x) \cdot g(x))' = f'(x) \cdot g(x) + f(x) \cdot g'(x)
$

Example:
If $f(x) = x^2$ and $g(x) = 3x$, then:

$
(f(x) \cdot g(x))' = 2x \cdot 3x + x^2 \cdot 3 = 6x^2 + 3x^2
$

### **Quotient Rule**
The derivative of a quotient of two functions is:

$
\left(\frac{f(x)}{g(x)}\right)' = \frac{f'(x) \cdot g(x) - f(x) \cdot g'(x)}{g(x)^2}
$

Example:
If $f(x) = x^2$ and $g(x) = 3x$, then:

$
\left(\frac{f(x)}{g(x)}\right)' = \frac{2x \cdot 3x - x^2 \cdot 3}{(3x)^2}
$

## 3. **Higher Order Derivatives**

The second derivative is the derivative of the first derivative and represents the rate of change of the rate of change. It helps to analyze the concavity of the function and determine whether it is concave up or concave down.

The notation for higher derivatives is:

- $f''(x)$ for the second derivative
- $f^{(n)}(x)$ for the nth derivative

Example:
If $f(x) = x^3$, the first and second derivatives are:

$
f'(x) = 3x^2
$

$
f''(x) = 6x
$

## 4. **Critical Points and Extrema**

To find the critical points of a function, we set the first derivative equal to zero:

$
f'(x) = 0
$

Critical points are where the function has a local maximum, local minimum, or inflection point.

To determine the nature of the critical points (whether they correspond to maxima or minima), we use the second derivative test. If $f''(x) > 0$, the function has a local minimum at that point. If $f''(x) < 0$, the function has a local maximum.

## 5. **Example: Derivative of a Polynomial Function**

Let's compute the derivative of the function:

$
f(x) = x^2 + 3x - 5
$

### **First Derivative**

Using the power rule, the derivative of $f(x) = x^2 + 3x - 5$ is:

$
f'(x) = 2x + 3
$

### **Second Derivative**

Taking the derivative of the first derivative:

$
f''(x) = 2
$

This means the function is concave up and has no local maxima or minima, just a straight line increase.



# Derivatives of Given Functions (Step-by-Step Explanation)

---

### 1. $y(x) = -3x + 3$  
 Use the **linear rule**. The derivative of $ax + b$ is $a$.  
**Derivative:**  
$$y'(x) = -3$$  

---

### 2. $y(x) = \pi x + \sin(1)$  
 Use the **constant rule**. The derivative of a constant is $0$, and the derivative of $\pi x$ is $\pi$.  
**Derivative:**  
$$y'(x) = \pi$$  

---

### 3. $y(x) = 4 + \sin(2)$  
 Use the **constant rule**. Both terms are constants, so their derivative is $0$.  
**Derivative:**  
$$y'(x) = 0$$  

---

### 4. $y(x) = 2x^3 - 3x^2 + 8x - 9$  
 Apply the **power rule** to each term. The derivative of $ax^n$ is $nax^{n-1}$.  
**Derivative:**  
$$y'(x) = 6x^2 - 6x + 8$$  

---

### 5. $y(x) = 6x^{1/3}$  
 Use the **power rule**. For fractional exponents, $d/dx(x^n) = nx^{n-1}$.  
**Derivative:**  
$$y'(x) = 2x^{-2/3}$$  

---

### 6. $y(x) = x$  
 The derivative of $x$ is $1$.  
**Derivative:**  
$$y'(x) = 1$$  

---

### 7. $y(x) = \cos(x) + \sin(x)$  
 Use the **trigonometric derivatives**. The derivative of $\cos(x)$ is $-\sin(x)$, and for $\sin(x)$, it is $\cos(x)$.  
**Derivative:**  
$$y'(x) = -\sin(x) + \cos(x)$$  

---

### 8. $y(x) = 2\sin(x)\cos(x)$  
 Apply the **product rule**: $d/dx(uv) = u'v + uv'$. Also, use the derivative of $\sin$ and $\cos$.  
**Derivative:**  
$$y'(x) = 2[\cos^2(x) - \sin^2(x)]$$  

---

### 9. $y(x) = x\sin(x)$  
 Use the **product rule**: $d/dx(uv) = u'v + uv'$.  
**Derivative:**  
$$y'(x) = \sin(x) + x\cos(x)$$  

---

### 10. $y(x) = (x+1)(x+1)$  
 Simplify as $(x+1)^2$, then apply the **chain rule** and **power rule**.  
**Derivative:**  
$$y'(x) = 2(x+1)$$  

---

### 11. $y(x) = \frac{x}{x+1}$  
 Use the **quotient rule**: $d/dx(u/v) = \frac{u'v - uv'}{v^2}$.  
**Derivative:**  
$$y'(x) = \frac{1}{(x+1)^2}$$  

---

### 12. $y(x) = (x+1)\exp(x)$  
 Use the **product rule**: $d/dx(uv) = u'v + uv'$. Also, use the derivative of $\exp(x)$, which is itself.  
**Derivative:**  
$$y'(x) = \exp(x) + (x+1)\exp(x)$$  

---

### 13. $y(x) = \sin(x^2)$  
 Use the **chain rule**. The derivative of $\sin(u)$ is $\cos(u)$, and then multiply by the derivative of $x^2$.  
**Derivative:**  
$$y'(x) = 2x\cos(x^2)$$  

---

### 14. $y(x) = \exp(-2x)$  
 Use the **chain rule**. The derivative of $\exp(u)$ is $\exp(u)$, multiplied by the derivative of $-2x$.  
**Derivative:**  
$$y'(x) = -2\exp(-2x)$$  

---

### 15. $y(x) = \frac{1}{\sin(x+1)}$  
 Rewrite as $[\sin(x+1)]^{-1}$ and use the **chain rule**.  
**Derivative:**  
$$y'(x) = -\frac{\cos(x+1)}{\sin^2(x+1)}$$  

---

### 16. $y(x) = \frac{2}{x+1}$  
 Use the **quotient rule** or rewrite as $2(x+1)^{-1}$ and apply the **power rule**.  
**Derivative:**  
$$y'(x) = -\frac{2}{(x+1)^2}$$  


#Example-2

## Proof:

$
\frac{d}{dx} \left( \ln(\sin(x)) \right) = \cot(x)
$

### Step 1: Apply the Chain Rule

$
\frac{d}{dx} \left( \ln(\sin(x)) \right) = \frac{1}{\sin(x)} \cdot \frac{d}{dx} (\sin(x))
$

### Step 2: Differentiate $\sin(x)$

$
\frac{d}{dx} (\sin(x)) = \cos(x)
$

### Step 3: Substitute and Simplify

$
\frac{d}{dx} \left( \ln(\sin(x)) \right) = \frac{1}{\sin(x)} \cdot \cos(x)
$

This simplifies to:

$
\frac{d}{dx} \left( \ln(\sin(x)) \right) = \frac{\cos(x)}{\sin(x)}
$

### Step 4: Recognize the Cotangent Function

$
\frac{\cos(x)}{\sin(x)} = \cot(x)
$

### Final Result:

$
\frac{d}{dx} \left( \ln(\sin(x)) \right) = \cot(x)
$
