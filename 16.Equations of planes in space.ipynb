# Equations of Planes in Space

A plane in three-dimensional space is defined by a point and a normal vector. There are several forms of the equation of a plane based on the given information.

## General Equation of a Plane

The general equation of a plane in three-dimensional space is:

$$
Ax + By + Cz + D = 0
$$

Where:
- $A$, $B$, and $C$ are the components of a normal vector to the plane, $\mathbf{n} = (A, B, C)$.
- $x$, $y$, and $z$ are the coordinates of any point on the plane.
- $D$ is a constant that determines the position of the plane relative to the origin.

### Vector Form of the Equation of a Plane

If we know:
- A point $P(x_0, y_0, z_0)$ on the plane.
- A normal vector $\mathbf{n} = (A, B, C)$.

The equation of the plane in vector form is:

$$
\mathbf{n} \cdot (\mathbf{r} - \mathbf{r_0}) = 0
$$

Where:
- $\mathbf{r}$ is the position vector of any point $(x, y, z)$ on the plane.
- $\mathbf{r_0}$ is the position vector of the given point $P(x_0, y_0, z_0)$.
- $\mathbf{n} \cdot (\mathbf{r} - \mathbf{r_0})$ represents the dot product of the normal vector and the vector from the point $P$ to $(x, y, z)$.

Expanding the equation gives:

$$
A(x - x_0) + B(y - y_0) + C(z - z_0) = 0
$$

This is the **point-normal form** of the equation of a plane.

## Standard Form of the Equation of a Plane

The **standard form** of the equation of a plane is derived from the point-normal form:

$$
Ax + By + Cz + D = 0
$$

Where:
- $A$, $B$, and $C$ are the components of the normal vector.
- $D$ is the constant determined by substituting the coordinates of a known point on the plane into the equation.

## Equation of a Plane Through Three Points

If we are given three points $P_1(x_1, y_1, z_1)$, $P_2(x_2, y_2, z_2)$, and $P_3(x_3, y_3, z_3)$ that do not lie on the same line, we can find the equation of the plane passing through these points.

### Steps:
1. Find two vectors $\mathbf{v_1}$ and $\mathbf{v_2}$ that lie on the plane:
   $$ \mathbf{v_1} = \mathbf{P_2} - \mathbf{P_1} $$
   $$ \mathbf{v_2} = \mathbf{P_3} - \mathbf{P_1} $$

2. Calculate the normal vector $\mathbf{n}$ using the cross product of $\mathbf{v_1}$ and $\mathbf{v_2}$:
   $$ \mathbf{n} = \mathbf{v_1} \times \mathbf{v_2} $$

3. Use the point-normal form of the equation of the plane:
   $$ \mathbf{n} \cdot (\mathbf{r} - \mathbf{r_0}) = 0 $$

   Where $\mathbf{r_0}$ is the position vector of one of the points, say $P_1$.

4. Expand the equation to get the standard form.

## Equation of a Plane Parallel to a Given Plane

If we are given a plane with the equation:

$$
Ax + By + Cz + D = 0
$$

and we are asked to find a plane parallel to it that passes through a point $(x_0, y_0, z_0)$, the equation of the new plane will have the same normal vector $\mathbf{n} = (A, B, C)$ and will be of the form:

$$
A(x - x_0) + B(y - y_0) + C(z - z_0) = 0
$$

This is the point-normal form of the plane.

## Equation of a Plane Perpendicular to a Given Plane

If we are given a plane with the equation:

$$
Ax + By + Cz + D = 0
$$

and we are asked to find a plane perpendicular to it that passes through a point $(x_0, y_0, z_0)$, the normal vector of the new plane must be parallel to the direction of the given plane's normal vector. Thus, the equation of the perpendicular plane is also of the form:

$$
A(x - x_0) + B(y - y_0) + C(z - z_0) = 0
$$

## Conclusion

The equation of a plane in space can be expressed in several forms depending on the given information:
1. **General form**: $Ax + By + Cz + D = 0$
2. **Point-normal form**: $A(x - x_0) + B(y - y_0) + C(z - z_0) = 0$
3. **Vector form**: $\mathbf{n} \cdot (\mathbf{r} - \mathbf{r_0}) = 0$
4. **Through three points**: Use cross product to find the normal vector.
5. **Parallel and perpendicular planes**: Derived from the point-normal form.

These forms are useful for describing planes in space and solving related geometric problems.


#Example-1

# Equation of the Plane Passing Through Points A, B, and C

Given points:
- A(1, 2, 3)
- B(3, 4, 5)
- C(2, 1, 4)

## Step 1: Find Two Vectors in the Plane
- $\mathbf{AB} = (3 - 1, 4 - 2, 5 - 3) = (2, 2, 2)$
- $\mathbf{AC} = (2 - 1, 1 - 2, 4 - 3) = (1, -1, 1)$

## Step 2: Find the Normal Vector
- $\mathbf{n} = \mathbf{AB} \times \mathbf{AC} = (4, 0, -4)$

## Step 3: Write the Equation of the Plane
Using point A(1, 2, 3) and normal vector $\mathbf{n} = (4, 0, -4)$:
$$
4(x - 1) + 0(y - 2) - 4(z - 3) = 0
$$
Simplifying:
$$
x - z + 2 = 0
$$

## Conclusion
The equation of the plane is:
$$
x - z + 2 = 0
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the points A, B, C
A = np.array([1, 2, 3])
B = np.array([3, 4, 5])
C = np.array([2, 1, 4])

# Vectors AB and AC
AB = B - A
AC = C - A

# Normal vector to the plane (cross product of AB and AC)
normal_vector = np.cross(AB, AC)

# Plane equation: Ax + By + Cz = D
A_plane, B_plane, C_plane = normal_vector
D = np.dot(normal_vector, A)

# Display the equation of the plane
print(f"The equation of the plane is: {A_plane}x + {B_plane}y + {C_plane}z = {D}")

# Create a grid for the plot
x = np.linspace(0, 4, 10)
y = np.linspace(0, 5, 10)
X, Y = np.meshgrid(x, y)

# Calculate Z using the plane equation
Z = (D - A_plane*X - B_plane*Y) / C_plane

# Plot the plane
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100)

# Plot the points A, B, C
ax.scatter([A[0], B[0], C[0]], [A[1], B[1], C[1]], [A[2], B[2], C[2]], color='r', s=50)

# Labels for the points
ax.text(A[0], A[1], A[2], 'A', color='r')
ax.text(B[0], B[1], B[2], 'B', color='r')
ax.text(C[0], C[1], C[2], 'C', color='r')

# Axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

# Equation of the Plane Passing Through Point A and Parallel to Another Plane

Given:
- Point A(1, 2, 3)
- Plane: $2x + 3y + 4z = 5$

## Step 1: Use the Normal Vector of the Given Plane

The normal vector of the given plane is the coefficient of $x$, $y$, and $z$:
$$
\mathbf{n} = (2, 3, 4)
$$

Since the new plane is parallel to the given plane, the normal vector of the new plane is also $(2, 3, 4)$.

## Step 2: Use Point A and Normal Vector in Point-Normal Form

Using point A(1, 2, 3) and normal vector $(2, 3, 4)$, the equation of the plane is:

$$
2(x - 1) + 3(y - 2) + 4(z - 3) = 0
$$

Simplifying:

$$
2x - 2 + 3y - 6 + 4z - 12 = 0
$$

$$
2x + 3y + 4z - 20 = 0
$$

## Conclusion
The equation of the plane is:
$$
2x + 3y + 4z = 20
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Given point A(1, 2, 3)
A = np.array([1, 2, 3])

# Normal vector of the plane
normal_vector = np.array([2, 3, 4])

# Calculate D using the point A
D = np.dot(normal_vector, A)

# Display the equation of the plane
print(f"The equation of the plane is: 2x + 3y + 4z = {D}")

# Create a grid for the plot
x = np.linspace(0, 4, 10)
y = np.linspace(0, 5, 10)
X, Y = np.meshgrid(x, y)

# Calculate Z using the plane equation
Z = (D - normal_vector[0]*X - normal_vector[1]*Y) / normal_vector[2]

# Plot the plane
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100)

# Plot the point A
ax.scatter(A[0], A[1], A[2], color='r', s=50)

# Label the point A
ax.text(A[0], A[1], A[2], 'A', color='r')

# Axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

$$
\mathbf{n} \cdot \left( (x, y, z) - (x_0, y_0, z_0) \right) = 0

Where $(x_0, y_0, z_0)$ is a point on the plane, and $\mathbf{n}$ is the normal vector.

2(x - 1) + 3(y - 2) + 4(z - 3) = 0

$$
2x - 2 + 3y - 6 + 4z - 12 = 0

$$
2x + 3y + 4z = 20

# Line of Intersection of Two Planes

Given the planes:
- $2x + 3y + 4z = 5$
- $3x + 4y + 2z = 6$

## Step 1: Find the Direction Vector of the Line

The direction vector of the line of intersection can be found by taking the cross product of the normal vectors of the planes.

- Normal vector of the first plane: $\mathbf{n_1} = (2, 3, 4)$
- Normal vector of the second plane: $\mathbf{n_2} = (3, 4, 2)$

The direction vector of the line is $\mathbf{v} = \mathbf{n_1} \times \mathbf{n_2}$:

$$
\mathbf{v} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ 2 & 3 & 4 \\ 3 & 4 & 2 \end{vmatrix}
$$

Calculating the determinant:

- For $\hat{i}$: $(3)(2) - (4)(4) = 6 - 16 = -10$
- For $\hat{j}$: $(2)(2) - (4)(3) = 4 - 12 = -8$
- For $\hat{k}$: $(2)(4) - (3)(3) = 8 - 9 = -1$

Thus, the direction vector is:

$$
\mathbf{v} = (-10, -8, -1)
$$

## Step 2: Find a Point on the Line

To find a point on the line, solve the system of equations by setting one variable (say $z = 0$) and solving for $x$ and $y$.

For $z = 0$:
- Plane 1: $2x + 3y = 5$
- Plane 2: $3x + 4y = 6$

Solve this system of equations.

From the first equation: $2x + 3y = 5$, solve for $x$:

$$
x = \frac{5 - 3y}{2}
$$

Substitute into the second equation:

$$
3\left(\frac{5 - 3y}{2}\right) + 4y = 6
$$

Simplify and solve for $y$:

$$
\frac{15 - 9y}{2} + 4y = 6
$$

Multiply through by 2:

$$
15 - 9y + 8y = 12
$$

Simplifying:

$$
15 - y = 12
$$

Thus, $y = 3$.

Substitute $y = 3$ into $2x + 3y = 5$:

$$
2x + 9 = 5 \Rightarrow 2x = -4 \Rightarrow x = -2
$$

So, the point on the line is $(-2, 3, 0)$.

## Step 3: Write the Parametric Equations of the Line

Using the point $(-2, 3, 0)$ and direction vector $(-10, -8, -1)$, the parametric equations of the line are:

$$
x = -2 - 10t
$$
$$
y = 3 - 8t
$$
$$
z = -t
$$

## Conclusion

The line of intersection of the two planes is given by the parametric equations:

$$
x = -2 - 10t, \quad y = 3 - 8t, \quad z = -t
$$


```
# Importing necessary libraries for 3D plotting
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the two planes:
# Plane 1: 2x + 3y + 4z = 5
# Plane 2: 3x + 4y + 2z = 6

# Create a grid for X and Y values for visualization
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

# Equation of Plane 1: 2x + 3y + 4z = 5 -> solve for Z
Z1 = (5 - 2*X - 3*Y) / 4

# Equation of Plane 2: 3x + 4y + 2z = 6 -> solve for Z
Z2 = (6 - 3*X - 4*Y) / 2

# Create plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the two planes
ax.plot_surface(X, Y, Z1, alpha=0.5, cmap='viridis', edgecolor='none')
ax.plot_surface(X, Y, Z2, alpha=0.5, cmap='plasma', edgecolor='none')

# Plot line of intersection of planes (a simple parametric approach)
t = np.linspace(-5, 5, 100)
x_line = t
y_line = t
z_line = (5 - 2*x_line - 3*y_line) / 4  # Using Plane 1 equation to find z

ax.plot(x_line, y_line, z_line, color='red', label="Line of Intersection")

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show plot
plt.legend()
plt.show()
```

# Equation of the Plane Passing Through Point A and Parallel to Vectors v1 and v2

Given:
- Point A(1, 2, 3)
- Vectors $\mathbf{v_1} = (1, 0, 1)$ and $\mathbf{v_2} = (0, 1, -1)$

## Step 1: Find the Normal Vector to the Plane

The normal vector $\mathbf{n}$ to the plane is the cross product of vectors $\mathbf{v_1}$ and $\mathbf{v_2}$:

$$
\mathbf{n} = \mathbf{v_1} \times \mathbf{v_2}
$$

Using the determinant to calculate the cross product:

$$
\mathbf{n} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ 1 & 0 & 1 \\ 0 & 1 & -1 \end{vmatrix}
$$

Calculating the determinant:

- For $\hat{i}$: $(0)(-1) - (1)(1) = -1$
- For $\hat{j}$: $(1)(-1) - (1)(0) = -1$
- For $\hat{k}$: $(1)(1) - (0)(0) = 1$

Thus, the normal vector is:

$$
\mathbf{n} = (-1, -1, 1)
$$

## Step 2: Use Point-Normal Form

The point-normal form of the equation of the plane is:

$$
\mathbf{n} \cdot \left( (x, y, z) - (x_0, y_0, z_0) \right) = 0
$$

Substitute point A(1, 2, 3) and normal vector $\mathbf{n} = (-1, -1, 1)$:

$$
-1(x - 1) - 1(y - 2) + 1(z - 3) = 0
$$

Simplify the equation:

$$
-(x - 1) - (y - 2) + (z - 3) = 0
$$

$$
-x + 1 - y + 2 + z - 3 = 0
$$

$$
-x - y + z = 0
$$

## Conclusion
The equation of the plane is:
$$
x + y - z = 0
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the point A and the vectors v1 and v2
A = np.array([1, 2, 3])
v1 = np.array([1, 0, 1])
v2 = np.array([0, 1, -1])

# Compute the cross product of v1 and v2
normal_vector = np.cross(v1, v2)

# The equation of the plane is:
# normal_vector . (r - A) = 0
# or normal_vector[0]*(x - A[0]) + normal_vector[1]*(y - A[1]) + normal_vector[2]*(z - A[2]) = 0

# Create a meshgrid for the x and y axes
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Calculate Z values from the plane equation
Z = (normal_vector[0] * (X - A[0]) + normal_vector[1] * (Y - A[1])) / -normal_vector[2] + A[2]

# Plot the plane
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100)

# Plot the point A
ax.scatter(A[0], A[1], A[2], color='r', s=100)

# Plot vectors v1 and v2 from point A
ax.quiver(A[0], A[1], A[2], v1[0], v1[1], v1[2], length=2, color='g')
ax.quiver(A[0], A[1], A[2], v2[0], v2[1], v2[2], length=2, color='b')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

# Finding Planes Parallel and Perpendicular to Given Plane

Given the plane:
$$
2x + 3y + 4z = 5
$$

## Step 1: Plane Parallel to the Given Plane

The normal vector of the given plane is $\mathbf{n} = (2, 3, 4)$. Any plane parallel to this plane will have the same normal vector.

To find a parallel plane, simply change the constant term. For example, choose a new constant term, say 10. The equation of a plane parallel to the given plane is:

$$
2x + 3y + 4z = 10
$$

Thus, the equation of a plane parallel to the given plane is:
$$
2x + 3y + 4z = 10
$$

## Step 2: Plane Perpendicular to the Given Plane

A plane perpendicular to the given plane will have a normal vector that is perpendicular to the normal vector of the given plane.

To find such a plane, we can use any vector that is perpendicular to $\mathbf{n} = (2, 3, 4)$. For example, take the vector $\mathbf{v} = (1, -1, 0)$, which is perpendicular to $\mathbf{n}$ because:

$$
2(1) + 3(-1) + 4(0) = 2 - 3 = -1 \neq 0
$$

The normal vector to the perpendicular plane will be $\mathbf{v} = (1, -1, 0)$. So, the equation of a plane perpendicular to the given plane passing through a point, say $(1, 1, 1)$, is:

$$
1(x - 1) - 1(y - 1) + 0(z - 1) = 0
$$

Simplifying:

$$
x - y = 0
$$

Thus, the equation of a plane perpendicular to the given plane is:
$$
x - y = 0
$$

## Conclusion

- A parallel plane: $2x + 3y + 4z = 10$
- A perpendicular plane: $x - y = 0$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the coefficients for the planes
normal_vector_1 = np.array([2, 3, 4])  # Normal vector for the given plane
normal_vector_2 = np.array([2, 3, 4])  # Same normal vector for the parallel plane
normal_vector_3 = np.array([-3, 2, 0])  # Normal vector for the perpendicular plane

# Create a meshgrid for the x and y axes
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Calculate Z values for the planes
Z1 = (5 - 2*X - 3*Y) / 4  # Plane 1: 2x + 3y + 4z = 5
Z2 = (10 - 2*X - 3*Y) / 4  # Parallel Plane: 2x + 3y + 4z = 10
Z3 = (0 - (-3*X + 2*Y)) / 0.0001  # Perpendicular plane: -3x + 2y = 0 (adjusting for non-zero Z)

# Plot the planes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the planes
ax.plot_surface(X, Y, Z1, alpha=0.5, rstride=100, cstride=100, color='r', label='Plane: 2x + 3y + 4z = 5')
ax.plot_surface(X, Y, Z2, alpha=0.5, rstride=100, cstride=100, color='b', label='Parallel Plane: 2x + 3y + 4z = 10')

# Plot the perpendicular plane (approximated by slicing Z values)
ax.plot_surface(X, Y, Z3, alpha=0.5, rstride=100, cstride=100, color='g', label='Perpendicular Plane: -3x + 2y = 0')

# Labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.legend()
plt.show()
```

# Distance from Point A to the Plane

Given:
- Plane: $2x + 3y + 4z = 5$
- Point A(1, 2, 3)

## Step 1: Formula for Distance from a Point to a Plane

The formula for the distance $d$ from a point $(x_0, y_0, z_0)$ to a plane $Ax + By + Cz + D = 0$ is:

$$
d = \frac{|Ax_0 + By_0 + Cz_0 + D|}{\sqrt{A^2 + B^2 + C^2}}
$$

Where:
- $(x_0, y_0, z_0)$ is the point,
- $Ax + By + Cz + D = 0$ is the equation of the plane.

## Step 2: Apply the Formula

The given plane equation is $2x + 3y + 4z = 5$. We rewrite it as:

$$
2x + 3y + 4z - 5 = 0
$$

Here, $A = 2$, $B = 3$, $C = 4$, and $D = -5$.

Now substitute point A(1, 2, 3) into the distance formula:

$$
d = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}}
$$

## Step 3: Simplify the Calculation

Simplify the numerator:

$$
2(1) + 3(2) + 4(3) - 5 = 2 + 6 + 12 - 5 = 15
$$

Now calculate the denominator:

$$
\sqrt{2^2 + 3^2 + 4^2} = \sqrt{4 + 9 + 16} = \sqrt{29}
$$

So, the distance is:

$$
d = \frac{|15|}{\sqrt{29}} = \frac{15}{\sqrt{29}}
$$

## Conclusion

The distance from point A(1, 2, 3) to the plane $2x + 3y + 4z = 5$ is:

$$
d = \frac{15}{\sqrt{29}}
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the point A and the plane coefficients
A_point = np.array([1, 2, 3])
plane_normal = np.array([2, 3, 4])  # Normal vector of the plane
D = -5  # The D coefficient of the plane equation: 2x + 3y + 4z - 5 = 0

# Compute the distance from point A to the plane
distance = abs(np.dot(plane_normal, A_point) + D) / np.linalg.norm(plane_normal)

# Output the distance
print(f"Distance from point A to the plane: {distance}")

# Create a meshgrid for plotting the plane
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Calculate Z values from the plane equation
Z = (5 - 2*X - 3*Y) / 4  # From the equation: 2x + 3y + 4z = 5

# Plot the plane and the point A
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of the plane
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100, color='r')

# Plot the point A
ax.scatter(A_point[0], A_point[1], A_point[2], color='g', s=100, label='Point A(1, 2, 3)')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot with the distance line
ax.text(A_point[0], A_point[1], A_point[2], f'Distance: {distance:.2f}', color='blue')

plt.show()
```

# Finding the Equation of the Plane

Given:
- Points of intersection with the coordinate axes:
  - $A(2, 0, 0)$
  - $B(0, 3, 0)$
  - $C(0, 0, 4)$

## Step 1: General Equation of the Plane

The general equation of a plane in 3D space is:

$$
\frac{x}{a} + \frac{y}{b} + \frac{z}{c} = 1
$$

Where $a$, $b$, and $c$ are the intercepts of the plane on the $x$, $y$, and $z$ axes, respectively.

## Step 2: Use Given Points to Determine the Intercepts

From the given points of intersection:
- Point $A(2, 0, 0)$ gives the $x$-intercept, so $a = 2$.
- Point $B(0, 3, 0)$ gives the $y$-intercept, so $b = 3$.
- Point $C(0, 0, 4)$ gives the $z$-intercept, so $c = 4$.

## Step 3: Write the Equation of the Plane

Substitute $a = 2$, $b = 3$, and $c = 4$ into the general equation:

$$
\frac{x}{2} + \frac{y}{3} + \frac{z}{4} = 1
$$

## Step 4: Clear the Fractions

To eliminate the fractions, multiply the entire equation by 12 (the least common multiple of 2, 3, and 4):

$$
6x + 4y + 3z = 12
$$

## Conclusion

The equation of the plane is:

$$
6x + 4y + 3z = 12
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the points A, B, C
A = np.array([2, 0, 0])
B = np.array([0, 3, 0])
C = np.array([0, 0, 4])

# Compute the vectors AB and AC
AB = B - A
AC = C - A

# Compute the normal vector to the plane (cross product of AB and AC)
normal_vector = np.cross(AB, AC)

# Create a meshgrid for the x and y axes
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Compute Z values from the plane equation
Z = (normal_vector[0] * (X - A[0]) + normal_vector[1] * (Y - A[1])) / -normal_vector[2] + A[2]

# Plot the plane and the points A, B, C
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of the plane
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100, color='r')

# Plot the points A, B, C
ax.scatter(A[0], A[1], A[2], color='g', s=100, label='Point A(2, 0, 0)')
ax.scatter(B[0], B[1], B[2], color='b', s=100, label='Point B(0, 3, 0)')
ax.scatter(C[0], C[1], C[2], color='y', s=100, label='Point C(0, 0, 4)')

# Labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.legend()
plt.show()
```

#Example-9

# Calculating the Angle Between Two Planes

Given:
- Plane 1: $x + y + z = 1$
- Plane 2: $x = 0$ (the $yz$-plane)

## Step 1: Normal Vectors of the Planes

The angle between two planes is determined by the angle between their normal vectors.

- The normal vector of Plane 1 ($x + y + z = 1$) is $\mathbf{n_1} = (1, 1, 1)$.
- The normal vector of Plane 2 ($x = 0$) is $\mathbf{n_2} = (1, 0, 0)$, as the $yz$-plane is perpendicular to the $x$-axis.

## Step 2: Formula for the Angle Between Two Vectors

The formula to find the angle $\theta$ between two vectors $\mathbf{n_1}$ and $\mathbf{n_2}$ is:

$$
\cos \theta = \frac{\mathbf{n_1} \cdot \mathbf{n_2}}{|\mathbf{n_1}| |\mathbf{n_2}|}
$$

Where:
- $\mathbf{n_1} \cdot \mathbf{n_2}$ is the dot product of the vectors,
- $|\mathbf{n_1}|$ and $|\mathbf{n_2}|$ are the magnitudes of the vectors.

## Step 3: Calculate the Dot Product

The dot product of $\mathbf{n_1} = (1, 1, 1)$ and $\mathbf{n_2} = (1, 0, 0)$ is:

$$
\mathbf{n_1} \cdot \mathbf{n_2} = (1)(1) + (1)(0) + (1)(0) = 1
$$

## Step 4: Calculate the Magnitudes of the Vectors

The magnitude of $\mathbf{n_1}$ is:

$$
|\mathbf{n_1}| = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3}
$$

The magnitude of $\mathbf{n_2}$ is:

$$
|\mathbf{n_2}| = \sqrt{1^2 + 0^2 + 0^2} = 1
$$

## Step 5: Calculate the Angle

Substitute the values into the formula:

$$
\cos \theta = \frac{1}{\sqrt{3} \times 1} = \frac{1}{\sqrt{3}}
$$

Thus, the angle $\theta$ is:

$$
\theta = \cos^{-1}\left(\frac{1}{\sqrt{3}}\right) \approx 54.74^\circ
$$

## Conclusion

The angle between the plane $x + y + z = 1$ and the plane $x = 0$ is approximately:

$$
54.74^\circ
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the normal vectors for the planes
n1 = np.array([1, 1, 1])  # Normal vector for plane 1: x + y + z = 1
n2 = np.array([1, 0, 0])  # Normal vector for plane 2: x = 0 (yz plane)

# Calculate the angle between the planes
cos_theta = np.dot(n1, n2) / (np.linalg.norm(n1) * np.linalg.norm(n2))
theta = np.arccos(cos_theta)  # Angle in radians
theta_deg = np.degrees(theta)  # Angle in degrees

# Print the angle
print(f"Angle between the planes: {theta_deg:.2f} degrees")

# Create a meshgrid for the x and y axes
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Plane 1 equation: x + y + z = 1 -> z = 1 - x - y
Z1 = 1 - X - Y

# Plane 2 is the yz-plane, so x = 0, we just plot the yz-plane part
Z2 = np.linspace(-5, 5, 10)
X2, Y2 = np.meshgrid(np.zeros_like(Z2), Z2)

# Plot the planes and the angle between them
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of plane 1 (x + y + z = 1)
ax.plot_surface(X, Y, Z1, alpha=0.5, color='r', rstride=100, cstride=100)

# Plot the yz-plane (x = 0)
ax.plot_surface(X2, Y2, np.zeros_like(X2), alpha=0.5, color='b', rstride=100, cstride=100)

# Labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.title(f"Angle between the planes: {theta_deg:.2f} degrees")
plt.show()
```

# Finding the Perpendicular Vector to the Plane

Given the plane equation:
$$
x + y + z = 1
$$

## Step 1: Normal Vector of the Plane

The vector perpendicular to the plane is the normal vector of the plane. For the plane equation $Ax + By + Cz = D$, the normal vector $\mathbf{n}$ is given by:

$$
\mathbf{n} = (A, B, C)
$$

In our case, the equation of the plane is $x + y + z = 1$, which can be written as:

$$
1x + 1y + 1z = 1
$$

So, the normal vector $\mathbf{n}$ is:

$$
\mathbf{n} = (1, 1, 1)
$$

## Conclusion

The vector perpendicular to the plane $x + y + z = 1$ is:

$$
\mathbf{n} = (1, 1, 1)
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the normal vector to the plane (x + y + z = 1)
normal_vector = np.array([1, 1, 1])

# Create a meshgrid for the x and y axes
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Plane equation: x + y + z = 1 -> z = 1 - x - y
Z = 1 - X - Y

# Plot the plane and the normal vector
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of the plane
ax.plot_surface(X, Y, Z, alpha=0.5, color='r', rstride=100, cstride=100)

# Define the origin (point on the plane) and plot the normal vector
origin = np.array([0, 0, 1])  # A point on the plane
ax.quiver(origin[0], origin[1], origin[2], normal_vector[0], normal_vector[1], normal_vector[2],
          color='g', length=3, normalize=True, label="Normal Vector")

# Labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot with the normal vector
plt.legend()
plt.show()
```
