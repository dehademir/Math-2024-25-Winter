```
import numpy as np
import matplotlib.pyplot as plt

A = np.array([-1, 2])
B = np.array([3, -2])
P = (1 * B + 3 * A) / (1 + 3)

a = np.array([3, 4])
a_OX = np.array([a[0], 0])
a_OY = np.array([0, a[1]])

b = np.array([1, 1])
proj_ab = (np.dot(a, b) / np.dot(b, b)) * b

proj_ba = (np.dot(b, a) / np.dot(a, a)) * a

plt.figure(figsize=(8, 8))

plt.quiver(A[0], A[1], P[0]-A[0], P[1]-A[1], angles='xy', scale_units='xy', scale=1, color='b', label="Dividing point P")
plt.quiver(A[0], A[1], B[0]-A[0], B[1]-A[1], angles='xy', scale_units='xy', scale=1, color='g', label="Line segment AB")
plt.text(P[0]+0.2, P[1]+0.2, 'P(0, 1)', fontsize=12)

plt.quiver(0, 0, a_OX[0], a_OX[1], angles='xy', scale_units='xy', scale=1, color='r', label="Projection of a onto OX")
plt.quiver(0, 0, a_OY[0], a_OY[1], angles='xy', scale_units='xy', scale=1, color='purple', label="Projection of a onto OY")
plt.quiver(0, 0, proj_ab[0], proj_ab[1], angles='xy', scale_units='xy', scale=1, color='orange', label="Projection of a onto b")
plt.quiver(0, 0, proj_ba[0], proj_ba[1], angles='xy', scale_units='xy', scale=1, color='pink', label="Projection of b onto a")

plt.xlim(-2, 4)
plt.ylim(-3, 5)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

print(f"Dividing point P: {P}")
print(f"Projection of a onto OX: {a_OX}")
print(f"Projection of a onto OY: {a_OY}")
print(f"Projection of a onto b: {proj_ab}")
print(f"Projection of b onto a: {proj_ba}")
```
