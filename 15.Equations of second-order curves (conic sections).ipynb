# Equations of Second-Order Curves (Conic Sections)

Conic sections are curves obtained by intersecting a plane with a double-napped cone. The primary conic sections are **parabola**, **ellipse**, **circle**, and **hyperbola**.

## General Second-Degree Equation
The general equation of a second-order curve is:

$$
Ax^2 + Bxy + Cy^2 + Dx + Ey + F = 0
$$

Where:
- $A$, $B$, $C$, $D$, $E$, and $F$ are constants.
- The type of conic depends on the values of $A$, $B$, and $C$.

---

## Classification of Conic Sections

### 1. **Circle**
- Condition: $A = C$ and $B = 0$.
- Equation:
  $$
  (x - h)^2 + (y - k)^2 = r^2
  $$
  Where $(h, k)$ is the center, and $r$ is the radius.

### 2. **Parabola**
- Condition: Either $A = 0$ or $C = 0$, but not both.
- Equation (Vertical axis):
  $$
  y = ax^2 + bx + c
  $$
- Equation (Horizontal axis):
  $$
  x = ay^2 + by + c
  $$

### 3. **Ellipse**
- Condition: $A \neq C$, $A > 0$, $C > 0$, and $B = 0$.
- Standard Equation:
  $$
  \frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1
  $$
  Where $a > b$ for a horizontal ellipse, and $b > a$ for a vertical ellipse.

### 4. **Hyperbola**
- Condition: $AC < 0$.
- Standard Equation (Horizontal hyperbola):
  $$
  \frac{(x - h)^2}{a^2} - \frac{(y - k)^2}{b^2} = 1
  $$
- Standard Equation (Vertical hyperbola):
  $$
  \frac{(y - k)^2}{a^2} - \frac{(x - h)^2}{b^2} = 1
  $$

---

## Degenerate Cases
Degenerate conics occur when the plane passes through the vertex of the cone:
1. **Point**: $Ax^2 + Bxy + Cy^2 = 0$ and all terms vanish.
2. **Line**: $Ax^2 + Bxy + Cy^2$ factors into linear terms.
3. **Pair of Lines**: $Ax^2 + Bxy + Cy^2 = 0$ factors into two distinct linear terms.

---

## Applications
- **Circles**: Used in geometry, design, and engineering.
- **Ellipses**: Found in planetary orbits and optics.
- **Parabolas**: Used in satellite dishes and projectile motion.
- **Hyperbolas**: Found in navigation and certain natural phenomena.

# Equation of a Circle

## Given:
- Center: $A(1, 2)$
- Radius: $r = 3$

## Standard Form of a Circle
The equation of a circle with center $(h, k)$ and radius $r$ is:

$$
(x - h)^2 + (y - k)^2 = r^2
$$

## Substituting the Values
Here, $h = 1$, $k = 2$, and $r = 3$. Substituting these into the formula:

$$
(x - 1)^2 + (y - 2)^2 = 3^2
$$

Simplify:

$$
(x - 1)^2 + (y - 2)^2 = 9
$$

## Final Equation
The equation of the circle is:

$$
(x - 1)^2 + (y - 2)^2 = 9
$$


```
import numpy as np
import matplotlib.pyplot as plt

# Define the center and radius
center = np.array([1, 2])
radius = 3

# Create theta values for the circle
theta = np.linspace(0, 2 * np.pi, 300)

# Parametric equations for the circle
x = center[0] + radius * np.cos(theta)
y = center[1] + radius * np.sin(theta)

# Plot the circle
plt.figure(figsize=(6, 6))
plt.plot(x, y, label="Circle: $(x - 1)^2 + (y - 2)^2 = 9$", color='b')

# Highlight the center of the circle
plt.scatter(center[0], center[1], color='r', zorder=5, label="Center A(1, 2)")

# Set equal scaling and limits
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-1, 5)
plt.ylim(-1, 5)

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Circle with center A(1, 2) and radius r = 3')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

# Equation of a Parabola

## Given:
- The parabola intersects the $x$-axis at $x = 2$ and $x = 4$.
- The parabola passes through the point $(3, 1)$.

## General Form of a Parabola
Since the parabola intersects the $x$-axis at $x = 2$ and $x = 4$, it can be written in factored form:

$$
y = a(x - 2)(x - 4)
$$

Here, $a$ is a constant to be determined.

---

## Step 1: Use the point $(3, 1)$
Substitute $x = 3$ and $y = 1$ into the equation to find $a$:

$$
1 = a(3 - 2)(3 - 4)
$$

Simplify:

$$
1 = a(1)(-1)
$$

$$
a = -1
$$

---

## Step 2: Write the Equation
Substitute $a = -1$ back into the equation:

$$
y = -1(x - 2)(x - 4)
$$

Expand the terms:

$$
y = -(x^2 - 6x + 8)
$$

Simplify:

$$
y = -x^2 + 6x - 8
$$

---

## Final Equation
The equation of the parabola is:

$$
y = -x^2 + 6x - 8
$$


```
import numpy as np
import matplotlib.pyplot as plt

# Define the roots of the parabola
x_roots = [2, 4]
# Define the point through which the parabola passes
x_point = 3
y_point = 1

# Equation of a parabola with roots at x=2 and x=4: y = a(x-2)(x-4)
# We can solve for 'a' using the point (3,1)
# Substituting x = 3, y = 1 into the equation: 1 = a(3-2)(3-4) -> a = -1

a = -1
x = np.linspace(0, 6, 400)
y = a * (x - 2) * (x - 4)

# Plot the parabola
plt.figure(figsize=(6, 6))
plt.plot(x, y, label="Parabola: $y = -1(x - 2)(x - 4)$", color='b')

# Mark the x-intercepts and the point (3, 1)
plt.scatter([2, 4], [0, 0], color='r', zorder=5, label="Roots: x=2, x=4")
plt.scatter(x_point, y_point, color='g', zorder=5, label="Point (3, 1)")

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Parabola passing through (3, 1) and intersecting the x-axis at x=2 and x=4')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

# Finding the Center of an Ellipse

## Given Equation:
$$
x^2 + 4y^2 - 4x - 16y + 16 = 0
$$

---

## Step 1: Rearrange the Terms
Group $x$ and $y$ terms together:

$$
(x^2 - 4x) + (4y^2 - 16y) = -16
$$

---

## Step 2: Complete the Square

### For $x$:
Take the coefficient of $x$, divide it by $2$, and square it:

$$
\left(\frac{-4}{2}\right)^2 = 4
$$

Add and subtract $4$ within the $x$ terms:

$$
x^2 - 4x = (x - 2)^2 - 4
$$

### For $y$:
Factor $4$ out of the $y$ terms, then complete the square:

$$
4(y^2 - 4y)
$$

Take the coefficient of $y$, divide it by $2$, and square it:

$$
\left(\frac{-4}{2}\right)^2 = 4
$$

Add and subtract $4$ within the $y$ terms:

$$
4(y^2 - 4y) = 4((y - 2)^2 - 4) = 4(y - 2)^2 - 16
$$

---

## Step 3: Rewrite the Equation
Substitute the completed squares back into the equation:

$$
(x - 2)^2 - 4 + 4(y - 2)^2 - 16 = -16
$$

Simplify:

$$
(x - 2)^2 + 4(y - 2)^2 = 4
$$

---

## Step 4: Identify the Center
The standard form of an ellipse is:

$$
\frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1
$$

Here, the center is $(h, k)$. Comparing with the equation:

$$
(x - 2)^2 + 4(y - 2)^2 = 4
$$

The center is:

$$
(2, 2)
$$

---

## Final Answer:
The center of the ellipse is:

$$
(2, 2)
$$


```
import numpy as np
import matplotlib.pyplot as plt

# Define the equation of the ellipse: x^2 + 4y^2 - 4x - 16y + 16 = 0
# Rearranging terms to get the standard form:
# (x - h)^2 / a^2 + (y - k)^2 / b^2 = 1

# Completing the square:
# For x: x^2 - 4x -> (x - 2)^2 - 4
# For y: 4y^2 - 16y -> 4(y - 2)^2 - 16
# So the equation becomes: (x - 2)^2 / 4 + (y - 2)^2 / 4 = 1

# Center of the ellipse is at (h, k) = (2, 2)

center = (2, 2)

# Create a plot to visualize the ellipse and its center
theta = np.linspace(0, 2 * np.pi, 400)

# Parametric equations for the ellipse
x = 2 + 2 * np.cos(theta)
y = 2 + 2 * np.sin(theta)

# Plot the ellipse
plt.figure(figsize=(6, 6))
plt.plot(x, y, label="Ellipse: $(x - 2)^2/4 + (y - 2)^2/4 = 1$", color='b')

# Mark the center of the ellipse
plt.scatter(center[0], center[1], color='r', zorder=5, label="Center (2, 2)")

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Ellipse with center at (2, 2)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()

# Set equal scaling for both axes
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.grid(True)
plt.show()
```

# Finding the Slope of the Tangent Line

## Problem:
We need to find the slope $m$ ($m > 0$) of the line:
$$
y = mx - 5
$$
that is tangent to the circle:
$$
x^2 + y^2 = 1
$$

---

## Step 1: Substituting the Line Equation into the Circle Equation
Substitute $y = mx - 5$ into $x^2 + y^2 = 1$:

$$
x^2 + (mx - 5)^2 = 1
$$

Expand $(mx - 5)^2$:

$$
x^2 + m^2x^2 - 10mx + 25 = 1
$$

Combine terms:

$$
(1 + m^2)x^2 - 10mx + 24 = 0
$$

---

## Step 2: Condition for Tangency
For the line to be tangent to the circle, the quadratic equation in $x$ must have exactly one solution. This happens when the discriminant $\Delta$ is zero:

$$
\Delta = b^2 - 4ac = 0
$$

Here, the quadratic equation is:

$$
(1 + m^2)x^2 - 10mx + 24 = 0
$$

- $a = 1 + m^2$
- $b = -10m$
- $c = 24$

Substitute into $\Delta = b^2 - 4ac$:

$$
(-10m)^2 - 4(1 + m^2)(24) = 0
$$

Simplify:

$$
100m^2 - 96(1 + m^2) = 0
$$

---

## Step 3: Solve for $m$
Expand $96(1 + m^2)$:

$$
100m^2 - 96 - 96m^2 = 0
$$

Simplify:

$$
4m^2 = 96
$$

Solve for $m^2$:

$$
m^2 = 24
$$

Take the positive square root (since $m > 0$):

$$
m = \sqrt{24} = 2\sqrt{6}
$$

---

## Final Answer:
The slope of the tangent line is:

$$
m = 2\sqrt{6}
$$


```
import numpy as np
import matplotlib.pyplot as plt

# The equation of the circle is x^2 + y^2 = 1

# We know the point of tangency lies on the circle, so we need to solve for m (slope)
# The line equation is y = mx - 5. The tangent line touches the circle at exactly one point.

# Let's calculate the slope of the tangent line using the condition that the line intersects the circle at only one point.
# The discriminant of the quadratic equation obtained by substituting y = mx - 5 into the circle equation must be zero.

# Step 1: Define the circle and line equation
# Substituting y = mx - 5 into x^2 + y^2 = 1 gives us:
# x^2 + (mx - 5)^2 = 1
# Expanding this:
# x^2 + m^2*x^2 - 10mx + 25 = 1
# (1 + m^2)x^2 - 10mx + 24 = 0

# Step 2: The discriminant of this quadratic equation must be zero for a tangent line (one solution)
# Discriminant = B^2 - 4AC, where A = (1 + m^2), B = -10m, and C = 24
# Set discriminant to zero to find the slope.

from sympy import symbols, Eq, solve

# Define variables
m = symbols('m')

# Define the discriminant condition
A = 1 + m**2
B = -10 * m
C = 24
discriminant = B**2 - 4*A*C

# Solve for m when the discriminant is zero
slope_solution = solve(Eq(discriminant, 0), m)

# We want the positive slope m > 0
m_positive = max(slope_solution)

# Now let's visualize the circle and the tangent line
theta = np.linspace(0, 2 * np.pi, 400)
x_circle = np.cos(theta)
y_circle = np.sin(theta)

# Equation of the line y = mx - 5
x_line = np.linspace(-2, 2, 400)
y_line = m_positive * x_line - 5

# Create the plot
plt.figure(figsize=(6, 6))

# Plot the circle
plt.plot(x_circle, y_circle, label="Circle: $x^2 + y^2 = 1$", color='b')

# Plot the tangent line
plt.plot(x_line, y_line, label=f"Tangent line: $y = {m_positive:.2f}x - 5$", color='g')

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Tangent Line to the Circle $x^2 + y^2 = 1$')

# Mark the point of tangency
point_of_tangency = (5 / (m_positive**2 + 1), 5 * m_positive / (m_positive**2 + 1) - 5)
plt.scatter(*point_of_tangency, color='r', zorder=5, label="Point of Tangency")

# Set equal scaling for both axes
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

# Finding Intersection Points

## Given Equations:
1. Hyperbola:
   $$
   x^2 - y^2 = 1
   $$

2. Ellipse:
   $$
   x^2 + 4y^2 = 6
   $$

---

## Step 1: Express $x^2$ from the Hyperbola Equation
From $x^2 - y^2 = 1$:

$$
x^2 = 1 + y^2
$$

---

## Step 2: Substitute $x^2$ into the Ellipse Equation
Substitute $x^2 = 1 + y^2$ into $x^2 + 4y^2 = 6$:

$$
(1 + y^2) + 4y^2 = 6
$$

Simplify:

$$
1 + y^2 + 4y^2 = 6
$$

$$
1 + 5y^2 = 6
$$

$$
5y^2 = 5
$$

$$
y^2 = 1
$$

---

## Step 3: Solve for $y$
$$
y = \pm 1
$$

---

## Step 4: Solve for $x$
Substitute $y = 1$ and $y = -1$ into $x^2 = 1 + y^2$:

- When $y = 1$:
  $$
  x^2 = 1 + (1)^2 = 2
  $$
  $$
  x = \pm \sqrt{2}
  $$

- When $y = -1$:
  $$
  x^2 = 1 + (-1)^2 = 2
  $$
  $$
  x = \pm \sqrt{2}
  $$

---

## Step 5: Intersection Points
The intersection points are:

$$
(\sqrt{2}, 1), \, (-\sqrt{2}, 1), \, (\sqrt{2}, -1), \, (-\sqrt{2}, -1)
$$


```
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve

# Define symbols
x, y = symbols('x y')

# Define the equations
hyperbola_eq = Eq(x**2 - y**2, 1)
ellipse_eq = Eq(x**2 + 4*y**2, 6)

# Solve the system of equations
intersection_points = solve([hyperbola_eq, ellipse_eq], (x, y))

# Extract intersection points
intersection_points = [(point[0].evalf(), point[1].evalf()) for point in intersection_points]

# Create the plot for visualization
theta = np.linspace(0, 2 * np.pi, 400)

# Parametric equations for the ellipse x^2 + 4y^2 = 6
x_ellipse = np.sqrt(6) * np.cos(theta)
y_ellipse = np.sqrt(6) / 2 * np.sin(theta)

# Parametric equations for the hyperbola x^2 - y^2 = 1
x_hyperbola_positive = np.linspace(-2, 2, 400)
y_hyperbola_positive = np.sqrt(x_hyperbola_positive**2 - 1)

x_hyperbola_negative = np.linspace(-2, 2, 400)
y_hyperbola_negative = -np.sqrt(x_hyperbola_negative**2 - 1)

# Create the plot
plt.figure(figsize=(6, 6))

# Plot the ellipse
plt.plot(x_ellipse, y_ellipse, label="Ellipse: $x^2 + 4y^2 = 6$", color='b')

# Plot the hyperbola
plt.plot(x_hyperbola_positive, y_hyperbola_positive, label="Hyperbola: $x^2 - y^2 = 1$", color='r')
plt.plot(x_hyperbola_negative, y_hyperbola_negative, color='r')

# Plot the intersection points
for point in intersection_points:
    plt.scatter(point[0], point[1], color='g', label=f"Intersection Point {point}")

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Intersection of Hyperbola and Ellipse')

# Set equal scaling for both axes
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()

# Display intersection points
print("Intersection Points:", intersection_points)
```

# Distance Between the Branches of a Hyperbola

## Given Hyperbola:
$$
x^2 - y^2 = 1
$$

---

## Step 1: General Form
The general form of a hyperbola centered at the origin is:
$$
\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1
$$

Here:
- $a^2 = 1$, so $a = 1$
- $b^2 = 1$, so $b = 1$

---

## Step 2: Distance Between the Branches
The distance between the branches of a hyperbola is determined by the separation along the conjugate axis (the $y$-axis in this case). The distance is twice the length of $b$, which is given as $2b$.

For this hyperbola:
$$
b = 1
$$

Thus, the distance between the branches is:
$$
2b = 2(1) = 2
$$

---

## Final Answer:
The distance between the branches of the hyperbola is:
$$
2
$$


```
import numpy as np
import matplotlib.pyplot as plt

# Parametric equations for the hyperbola x^2 - y^2 = 1
x_hyperbola_positive = np.linspace(-2, 2, 400)
y_hyperbola_positive = np.sqrt(x_hyperbola_positive**2 - 1)

x_hyperbola_negative = np.linspace(-2, 2, 400)
y_hyperbola_negative = -np.sqrt(x_hyperbola_negative**2 - 1)

# Create the plot
plt.figure(figsize=(6, 6))

# Plot the hyperbola
plt.plot(x_hyperbola_positive, y_hyperbola_positive, label="Hyperbola: $x^2 - y^2 = 1$", color='r')
plt.plot(x_hyperbola_negative, y_hyperbola_negative, color='r')

# Mark the distance between branches at a specific point (x = 0)
distance = 2 * np.sqrt(0**2 - 1)  # Distance formula between branches along y-axis at x = 0
plt.annotate(f"Distance = {distance:.2f}", xy=(0, 0), xytext=(1, 1),
             arrowprops=dict(facecolor='black', arrowstyle="->"))

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Hyperbola: $x^2 - y^2 = 1$')

# Set equal scaling for both axes
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.show()
```
