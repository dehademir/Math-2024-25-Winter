# Inverse of a Matrix from the Formula

#### 1. Introduction
The **inverse of a matrix** $A$, denoted as $A^{-1}$, is a matrix that, when multiplied by $A$, gives the identity matrix:
$$
A \times A^{-1} = I
$$
where $I$ is the identity matrix. Not all matrices have an inverse. A matrix has an inverse only if it is **square** (same number of rows and columns) and its **determinant is non-zero**.

#### 2. Formula for the Inverse of a 2x2 Matrix
For a $2 \times 2$ matrix:
$$
A = \begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
$$
The inverse of $A$ is given by:
$$
A^{-1} = \frac{1}{\text{det}(A)} \begin{pmatrix}
d & -b \\
-c & a
\end{pmatrix}
$$
where $\text{det}(A) = ad - bc$.

##### Conditions:
- The matrix $A$ must be invertible, meaning $\text{det}(A) \neq 0$.
- If $\text{det}(A) = 0$, the matrix is **singular** and does not have an inverse.

#### 3. Formula for the Inverse of a 3x3 Matrix
For a $3 \times 3$ matrix:
$$
A = \begin{pmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{pmatrix}
$$
The inverse of $A$ is given by the formula:
$$
A^{-1} = \frac{1}{\text{det}(A)} \begin{pmatrix}
\text{det}(A_{11}) & -\text{det}(A_{12}) & \text{det}(A_{13}) \\
-\text{det}(A_{21}) & \text{det}(A_{22}) & -\text{det}(A_{23}) \\
\text{det}(A_{31}) & -\text{det}(A_{32}) & \text{det}(A_{33})
\end{pmatrix}
$$
where $\text{det}(A)$ is the determinant of matrix $A$, and each of $A_{ij}$ is the **minor** of the matrix $A$, obtained by deleting the $i$-th row and $j$-th column.

##### Steps to Find the Inverse:
1. **Find the determinant** of the matrix $A$.
2. If $\text{det}(A) = 0$, the matrix does not have an inverse.
3. **Find the minors** for each element in the matrix.
4. **Form the cofactor matrix** by applying signs ($+$ or $-$) to the minors.
5. **Transpose** the cofactor matrix to get the adjugate matrix.
6. Multiply the adjugate matrix by $\frac{1}{\text{det}(A)}$ to get the inverse matrix.

#### 4. Example for a $2 \times 2$ Matrix
Let’s find the inverse of the matrix:
$$
A = \begin{pmatrix}
4 & 7 \\
2 & 6
\end{pmatrix}
$$
First, calculate the determinant:
$$
\text{det}(A) = (4)(6) - (7)(2) = 24 - 14 = 10
$$
Since $\text{det}(A) = 10 \neq 0$, the matrix is invertible. Using the formula for a $2 \times 2$ matrix:
$$
A^{-1} = \frac{1}{\text{det}(A)} \begin{pmatrix}
6 & -7 \\
-2 & 4
\end{pmatrix}
$$
$$
A^{-1} = \frac{1}{10} \begin{pmatrix}
6 & -7 \\
-2 & 4
\end{pmatrix}
$$
Thus, the inverse of $A$ is:
$$
A^{-1} = \begin{pmatrix}
\frac{3}{5} & -\frac{7}{10} \\
-\frac{1}{5} & \frac{2}{5}
\end{pmatrix}
$$

#### 5. Example for a $3 \times 3$ Matrix
Let’s find the inverse of the matrix:
$$
A = \begin{pmatrix}
1 & 2 & 3 \\
0 & 1 & 4 \\
5 & 6 & 0
\end{pmatrix}
$$
First, calculate the determinant of $A$:
$$
\text{det}(A) = 1 \times \begin{vmatrix} 1 & 4 \\ 6 & 0 \end{vmatrix} - 2 \times \begin{vmatrix} 0 & 4 \\ 5 & 0 \end{vmatrix} + 3 \times \begin{vmatrix} 0 & 1 \\ 5 & 6 \end{vmatrix}
$$
$$
\text{det}(A) = 1 \times (1 \times 0 - 4 \times 6) - 2 \times (0 \times 0 - 4 \times 5) + 3 \times (0 \times 6 - 1 \times 5)
$$
$$
\text{det}(A) = 1 \times (-24) - 2 \times (-20) + 3 \times (-5)
$$
$$
\text{det}(A) = -24 + 40 - 15 = 1
$$
Since $\text{det}(A) = 1$, the matrix is invertible.

Now, find the minors, cofactors, and adjugate matrix. Finally, the inverse is:
$$
A^{-1} = \begin{pmatrix}
\text{det}(A_{11}) & -\text{det}(A_{12}) & \text{det}(A_{13}) \\
-\text{det}(A_{21}) & \text{det}(A_{22}) & -\text{det}(A_{23}) \\
\text{det}(A_{31}) & -\text{det}(A_{32}) & \text{det}(A_{33})
\end{pmatrix}
$$

#### 6. Conclusion
- To find the inverse of a matrix, first calculate its determinant.
- If the determinant is non-zero, use the formula for the inverse to compute it.
- For $2 \times 2$ matrices, the formula is straightforward, while for larger matrices, the process involves finding minors, cofactors, and the adjugate matrix before multiplying by $\frac{1}{\text{det}(A)}$.


#Example-1

### Finding the Inverse of Matrix $A$

Given matrix $A$:
$$
A = \begin{pmatrix}
2 & 0 & 1 \\
0 & 1 & 0 \\
1 & 2 & 0
\end{pmatrix}
$$

#### Step 1: Calculate the Determinant of $A$
The determinant of a $3 \times 3$ matrix is:
$$
\text{det}(A) = 2 \times \begin{vmatrix} 1 & 0 \\ 2 & 0 \end{vmatrix} - 0 + 1 \times \begin{vmatrix} 0 & 1 \\ 1 & 2 \end{vmatrix}
$$
Calculating the minors:
$$
\text{det}(A) = 2 \times (1 \times 0 - 0 \times 2) - 0 + 1 \times (0 \times 2 - 1 \times 1)
$$
$$
\text{det}(A) = 2 \times 0 - 0 + 1 \times (-1) = -1
$$

Since $\text{det}(A) = -1 \neq 0$, the matrix is invertible.

#### Step 2: Find the Cofactor Matrix of $A$
The cofactor matrix $C$ is found by calculating the minors and applying signs:
$$
C = \begin{pmatrix}
0 & 0 & -1 \\
2 & -1 & -4 \\
-1 & 1 & 2
\end{pmatrix}
$$

#### Step 3: Find the Adjugate Matrix
The adjugate matrix is the transpose of the cofactor matrix:
$$
\text{adj}(A) = C^T = \begin{pmatrix}
0 & 2 & -1 \\
0 & -1 & 1 \\
-1 & -4 & 2
\end{pmatrix}
$$

#### Step 4: Calculate the Inverse of $A$
The inverse is given by:
$$
A^{-1} = \frac{1}{\text{det}(A)} \times \text{adj}(A) = \frac{1}{-1} \times \begin{pmatrix}
0 & 2 & -1 \\
0 & -1 & 1 \\
-1 & -4 & 2
\end{pmatrix}
$$
$$
A^{-1} = \begin{pmatrix}
0 & 2 & -1 \\
0 & -1 & 1 \\
-1 & -4 & 2
\end{pmatrix}
$$

#### Step 5: Verify the Inverse
To verify the result, we multiply $A$ by $A^{-1}$:
$$
A \times A^{-1} = \begin{pmatrix}
2 & 0 & 1 \\
0 & 1 & 0 \\
1 & 2 & 0
\end{pmatrix}
\times \begin{pmatrix}
0 & 2 & -1 \\
0 & -1 & 1 \\
-1 & -4 & 2
\end{pmatrix}
$$
The result should be the identity matrix $I$:
$$
A \times A^{-1} = I = \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{pmatrix}
$$

Thus, the inverse is correct.

### Determining the Rank of Matrix $B$

Given matrix $B$:
$$
B = \begin{pmatrix}
4 & -3 & 7 \\
-1 & 6 & 3 \\
2 & 9 & 1
\end{pmatrix}
$$

#### Step 1: Write the Augmented Matrix (if applicable)
For determining the rank, we will apply row operations to reduce the matrix to row echelon form (REF).

#### Step 2: Perform Row Operations to Reach Row Echelon Form

Start with matrix $B$:
$$
B = \begin{pmatrix}
4 & -3 & 7 \\
-1 & 6 & 3 \\
2 & 9 & 1
\end{pmatrix}
$$

1. **Row 1:** Divide the first row by 4 to make the leading entry 1:
   $$ R_1 \rightarrow \frac{1}{4} R_1 = \begin{pmatrix} 1 & -\frac{3}{4} & \frac{7}{4} \end{pmatrix} $$

   New matrix:
   $$
   \begin{pmatrix}
   1 & -\frac{3}{4} & \frac{7}{4} \\
   -1 & 6 & 3 \\
   2 & 9 & 1
   \end{pmatrix}
   $$

2. **Row 2:** Add Row 1 to Row 2 to eliminate the first column entry in Row 2:
   $$ R_2 \rightarrow R_2 + R_1 $$

   New matrix:
   $$
   \begin{pmatrix}
   1 & -\frac{3}{4} & \frac{7}{4} \\
   0 & \frac{21}{4} & \frac{19}{4} \\
   2 & 9 & 1
   \end{pmatrix}
   $$

3. **Row 3:** Subtract 2 times Row 1 from Row 3 to eliminate the first column entry in Row 3:
   $$ R_3 \rightarrow R_3 - 2R_1 $$

   New matrix:
   $$
   \begin{pmatrix}
   1 & -\frac{3}{4} & \frac{7}{4} \\
   0 & \frac{21}{4} & \frac{19}{4} \\
   0 & \frac{39}{4} & -\frac{5}{2}
   \end{pmatrix}
   $$

4. **Row 2:** Multiply Row 2 by $\frac{4}{21}$ to make the leading entry in Row 2 equal to 1:
   $$ R_2 \rightarrow \frac{4}{21} R_2 $$

   New matrix:
   $$
   \begin{pmatrix}
   1 & -\frac{3}{4} & \frac{7}{4} \\
   0 & 1 & \frac{19}{21} \\
   0 & \frac{39}{4} & -\frac{5}{2}
   \end{pmatrix}
   $$

5. **Row 3:** Subtract $\frac{39}{4}$ times Row 2 from Row 3 to eliminate the second column entry in Row 3:
   $$ R_3 \rightarrow R_3 - \frac{39}{4} R_2 $$

   New matrix:
   $$
   \begin{pmatrix}
   1 & -\frac{3}{4} & \frac{7}{4} \\
   0 & 1 & \frac{19}{21} \\
   0 & 0 & -\frac{49}{21}
   \end{pmatrix}
   $$

#### Step 3: Count the Non-Zero Rows
The matrix is now in row echelon form, and we can observe that there are 3 non-zero rows. Therefore, the rank of matrix $B$ is:

$$
\text{rank}(B) = 3
$$
