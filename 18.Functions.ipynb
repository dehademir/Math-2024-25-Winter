# Functions

A **function** is a relation between a set of inputs (domain) and a set of possible outputs (codomain) such that each input is related to exactly one output.

## Definition:
A function $f$ from set $A$ to set $B$ is denoted as:

$$
f: A \to B
$$

Where for each element $x \in A$, there is a corresponding element $f(x) \in B$.

## Components of a Function:
1. **Domain**: The set of all possible input values for the function.
2. **Codomain**: The set of all possible output values.
3. **Range**: The actual outputs of the function, which may be a subset of the codomain.
4. **Function Rule**: The operation that defines how each input is mapped to an output.

## Notation:
For a function $f$, if $x$ is in the domain, the corresponding output is written as $f(x)$.

### Example:
Let $f(x) = x^2 + 1$. The domain could be all real numbers, and the function maps each $x$ to $x^2 + 1$.

## Types of Functions:
### 1. **One-to-One (Injective)**:
A function is **injective** if different inputs map to different outputs. In other words, if $f(x_1) = f(x_2)$, then $x_1 = x_2$.

### 2. **Onto (Surjective)**:
A function is **surjective** if for every element in the codomain, there is at least one element in the domain that maps to it. In other words, the range of the function is the entire codomain.

### 3. **Bijective**:
A function is **bijective** if it is both injective and surjective. This means that it is a one-to-one correspondence between the domain and codomain.

## Function Operations:
Functions can be combined or manipulated using various operations.

### 1. **Addition**:
If $f$ and $g$ are functions, the sum of $f$ and $g$ is defined as:
$$
(f + g)(x) = f(x) + g(x)
$$

### 2. **Multiplication**:
The product of $f$ and $g$ is defined as:
$$
(f \cdot g)(x) = f(x) \cdot g(x)
$$

### 3. **Composition**:
The composition of functions $f$ and $g$ is defined as:
$$
(f \circ g)(x) = f(g(x))
$$

### 4. **Inverse**:
A function $f$ has an **inverse** if there exists a function $f^{-1}$ such that:
$$
f^{-1}(f(x)) = x \quad \text{and} \quad f(f^{-1}(x)) = x
$$
This is only true if $f$ is bijective.

## Graph of a Function:
The graph of a function is a visual representation of how the inputs are mapped to outputs. The graph of a function is plotted on a coordinate plane, where the $x$-axis represents the inputs and the $y$-axis represents the outputs.

### Example of a Graph:
For $f(x) = x^2$, the graph is a parabola opening upwards.

## Domain and Range:
- **Domain**: The set of all possible input values for the function.
- **Range**: The set of all possible output values.

### Example:
For $f(x) = \sqrt{x}$, the domain is $x \geq 0$, because the square root of a negative number is not a real number. The range is $y \geq 0$.

## Special Functions:
### 1. **Linear Function**:
A linear function is of the form:
$$
f(x) = ax + b
$$
Where $a$ and $b$ are constants.

### 2. **Quadratic Function**:
A quadratic function is of the form:
$$
f(x) = ax^2 + bx + c
$$

### 3. **Exponential Function**:
An exponential function is of the form:
$$
f(x) = a^x
$$
Where $a$ is a constant and $a > 0$.

### 4. **Logarithmic Function**:
The logarithmic function is the inverse of the exponential function:
$$
f(x) = \log_a(x)
$$
Where $a$ is the base of the logarithm.

### 5. **Trigonometric Functions**:
Common trigonometric functions include:
- $f(x) = \sin(x)$
- $f(x) = \cos(x)$
- $f(x) = \tan(x)$

These functions relate the angles of a right triangle to the ratios of its sides.

## Continuity and Limits:
A function is **continuous** at a point if:
$$
\lim_{x \to c} f(x) = f(c)
$$

This means that the function does not have any jumps, holes, or breaks at that point.



## Conclusion:
Functions are fundamental concepts in mathematics, representing the relationship between inputs and outputs. They are used in many areas of mathematics and applied fields such as physics, economics, and engineering.

To find the inverse function of $f(x) = x - 1$:

1. Replace $f(x)$ with $y$:
   $$
   y = x - 1
   $$

2. Swap $x$ and $y$:
   $$
   x = y - 1
   $$

3. Solve for $y$:
   $$
   y = x + 1
   $$

Thus, the inverse function is:
$$
f^{-1}(x) = x + 1
$$
